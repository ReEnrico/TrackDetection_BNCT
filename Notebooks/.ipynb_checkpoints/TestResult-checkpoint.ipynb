```python
import glob
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
import os
```

```python
def UploadMainFile(name_array):
    image_array = []
    text_array = []
    lenght_array = []
    for name in name_array:
        img = io.imread("..\img_PseudoTracks\BW\{}.jpg".format(name))
        image_array.append(img)
        txt = np.loadtxt(os.path.join("..", "img_PseudoTracks", "blob_{}_coord.txt".format(name)))
        text_array.append(txt)
        lnght = len(txt)
        lenght_array.append(lnght)
    return image_array, text_array, lenght_array

def UploadResultFile(name_array):
    result_array = []
    for indx in range(len(name_array)):
        log = np.loadtxt("..\Test\Text\Log_Result\Result\{}_result.txt".format(names[indx]))
        dog = np.loadtxt("..\Test\Text\Dog_Result\Result\{}_result.txt".format(names[indx]))
        doh = np.loadtxt("..\Test\Text\Doh_Result\Result\{}_result.txt".format(names[indx]))
        result_array.append([log, dog, doh])
    return result_array

def SimplePlot(barray, image, name_img, number):
    fig, axs = plt.subplots(2, 2, constrained_layout=True, figsize=(10,6))
    fig.suptitle("Image {}, true tracks {}".format(name_img, number))

    axs[0, 0].plot(barray[0][:, 0], barray[0][:, 1])
    axs[0, 0].axhline(y=number, xmin=0, xmax=1, color="red")
    axs[0, 0].set_xlabel("Threshold")
    axs[0, 0].set_ylabel("#blobs")
    #axs[0, 0].set_yscale("log")
    axs[0, 0].set_title("Log")
  
    axs[0, 1].plot(barray[1][:, 0], barray[1][:, 1])
    axs[0, 1].axhline(y=number, xmin=0, xmax=1, color="red")
    axs[0, 1].set_xlabel("Threshold")
    axs[0, 1].set_ylabel("#blobs")
    #axs[0, 1].set_yscale("log")
    axs[0, 1].set_title("Dog")
  
    axs[1, 0].plot(barray[2][:, 0], barray[2][:, 1])
    axs[1, 0].axhline(y=number, xmin=0, xmax=1, color="red")
    axs[1, 0].set_xlabel("Threshold")
    axs[1, 0].set_ylabel("#blobs")
    #axs[1, 0].set_yscale("log")
    axs[1, 0].set_title("Doh")
  
    axs[1, 1].imshow(image, cmap="gray")
    axs[1, 1].set_title("Image displayed")

    plt.show()
    plt.close()

    
def GetFileName(namesarray, dirname):
    log_file = []
    dog_file = []
    doh_file = []
    for name in namesarray:
        log = glob.glob("..\Test\Text\Log_{dirname}\{name}_*.txt".format(name=name, dirname=dirname))
        log.sort()
        log_file.append(log)
    
        dog = glob.glob("..\Test\Text\Dog_{dirname}\{name}_*.txt".format(name=name, dirname=dirname))
        dog.sort()
        dog_file.append(dog)
    
        doh = glob.glob("..\Test\Text\Doh_{dirname}\{name}_*.txt".format(name=name, dirname=dirname))
        doh.sort()
        doh_file.append(doh)

    return log_file, dog_file, doh_file


def GetTracks(namesarray):
    track_array = []
    for i in range(len(namesarray)):
        tmp = []
        for j in range(len(namesarray[i])):
            dmp = np.loadtxt(namesarray[i][j])
            tmp.append(dmp)
        track_array.append(tmp)
    return track_array


def ImageCheck(barray, larray, narray):
    complete_image = []
    missing_image = []
    for i in range(len(narray)):
        if len(barray[i][0]) < larray[i]:
            missing_image.append(narray[i])
        elif len(barray[i][0]) >= larray[i]:
            complete_image.append(narray[i])
    return complete_image, missing_image


def GetText(img_missing):
    img_numb = []
    for num in img_missing:
        img_numb.append(names.index(num))  

    text_cluster = []
    for indx in range(len(img_missing)):
        text_cluster.append(texts[img_numb[indx]])
    
    return img_numb, text_cluster

def FindMissingBlobs(algo_array, img_missing, name_array):
    index, true_text = GetText(img_missing)
    missing_tracks = np.zeros((len(index), 5), dtype=object)
    for i in range(len(index)):
        correct_blob = np.zeros((len(true_text[i]), 4), dtype=object)
        for k in range(len(true_text[i])):
            x = true_text[i][k][1]
            y = true_text[i][k][0]
            r = true_text[i][k][2]
            color = true_text[i][k][3]
            for j in range(len(algo_array[index[i]][0])):
                d2 = (x - algo_array[index[i]][0][j][1])*(x - algo_array[index[i]][0][j][1]) + (y - algo_array[index[i]][0][j][0])*(y - algo_array[index[i]][0][j][0])
                if d2 <= r*r:
                    correct_blob[k] = np.array([y, x, r, color], dtype=object)
        zero_row, zero_column = np.where(correct_blob == 0)
        if len(zero_row) != 0:
            missing_tracks[i] = np.array([name_array[index[i]], true_text[i][zero_row[0]][0], true_text[i][zero_row[0]][1], true_text[i][zero_row[0]][2], true_text[i][zero_row[0]][3]], dtype=object)
        else:
            missing_tracks[i] = np.array([name_array[index[i]], 0, 0, 0, 0], dtype=object)
        
    return missing_tracks



def PlotMissImage(image, track_alg, point, name, label):
    if point[1] != 0:
        fig, axs = plt.subplots(1, 2, constrained_layout=True, figsize = (12, 8))
        fig.suptitle("Image {} Threshold 0.0131".format(name))
    
        axs[0].imshow(image, cmap='gray')
        #ax.scatter(track_alg[:,1], track_alg[:,0], s=track_alg[:,2], alpha=0.5, label=label, color="red")
        axs[0].scatter(point[2], point[1], s=point[3]*3, alpha=0.5, label="Missed", color="green") 
        
        axs[1].imshow(image, cmap='gray')
        
        plt.legend()
        plt.show()
        plt.close()
    
    else: 
        fig, ax = plt.subplots(1, 1, figsize = (12, 8))
        fig.suptitle("Image {} Threshold 0.0131".format(name))
    
        ax.imshow(image, cmap='gray')
        #ax.scatter(track_alg[:,1], track_alg[:,0], s=track_alg[:,2], alpha=0.5, label=label, color="red")
        #ax.scatter(point[2], point[1], s=point[3]*3, alpha=0.5, label="Missed", color="green") 
    
        #plt.legend()
        plt.show()
        plt.close()

        
def MatchBlob(true_array, algo_array, img_name, file_name, len_array, label):
    found_blob = np.zeros((len(img_name), 7), dtype=object)
    for i in range(len(img_name)):
        darkest_blob = min(true_array[i][:, 3])
        if len(file_name[i]) >= 1:
            counter = 0
            thresh = file_name[i][-1][-10:-4]
            color_array = []
            for k in range(len(true_array[i])):
                x = true_array[i][k][1]
                y = true_array[i][k][0]
                r = true_array[i][k][2]
                color = true_array[i][k][3]
                if np.isscalar(algo_array[i][-1][0]) == False:
                    for j in range(len(algo_array[i][-1])):
                        d2 = (x - algo_array[i][-1][j][1])*(x - algo_array[i][-1][j][1]) + (y - algo_array[i][-1][j][0])*(y - algo_array[i][-1][j][0])
                        if d2 <= r*r:
                            counter += 1
                            color_array.append(color)
                else:
                    d2 = (x - algo_array[i][-1][1])*(x - algo_array[i][-1][1]) + (y - algo_array[i][-1][0])*(y - algo_array[i][-1][0])
                    if d2 <= r*r:
                        counter += 1
                        color_array.append(color)
            found_blob[i] = np.array([img_name[i], len_array[i], darkest_blob, thresh, counter, min(color_array), max(color_array)], dtype=object)
            
            #np.savetxt("{}_Blob.txt".format(label), found_blob, fmt=['%3c', '%3i', '%.2f', '%.4f', '%3i', '%.2f', '%.2f'], header="{name:3s} {leng:2s} {tmcolor:5s} {num:2s} {mincol:5s} {maxcol}".format(name="name", leng="#blob", tmcolor="darkest", num="count", mincol="darkestfound", maxcol="brightestfound"))
    
            
        else:
            darkest_blob = min(true_array[i][:, 3])
            found_blob[i] = np.array([img_name[i], len_array[i], darkest_blob, 0, 0, 0, 0], dtype=object)
    
    np.save("{}_Blob.txt".format(label), found_blob)
    return(found_blob)



def BlobPlot(image, log_array, doh_array):
    if np.isscalar(log_array[0]) == False:
        fig, ax = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4), sharex=True, sharey=True)

        ax[0].imshow(image, cmap='gray')
        for j in range(len(log_array)):
            y = log_array[j][0]
            x = log_array[j][1]
            r = log_array[j][2]
            c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
            ax[0].add_patch(c)  
        ax[0].set_axis_off()

        ax[1].imshow(image, cmap='gray')
        for j in range(len(doh_array)):
            y = doh_array[j][0]
            x = doh_array[j][1]
            r = doh_array[j][2]
            c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
            ax[1].add_patch(c)
        ax[1].set_axis_off()
    
    else:
        fig, ax = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4), sharex=True, sharey=True)

        ax[0].imshow(image, cmap='gray')
        y = log_array[0]
        x = log_array[1]
        r = log_array[2]
        c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
        ax[0].add_patch(c)  
        ax[0].set_axis_off()

        ax[1].imshow(image, cmap='gray')
        y = doh_array[0]
        x = doh_array[1]
        r = doh_array[2]
        c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
        ax[1].add_patch(c)
        ax[1].set_axis_off()
        
    
    plt.show()
    plt.close()
    
    
    
    
def BlobDog(image, dog_array):
    if np.isscalar(dog_array[0]) == False:
        fig, ax = plt.subplots(1, 1, constrained_layout=True, figsize=(12, 4), sharex=True, sharey=True)

        ax.imshow(image, cmap='gray')
        for j in range(len(dog_array)):
            y = dog_array[j][0]
            x = dog_array[j][1]
            r = dog_array[j][2]
            c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
            ax.add_patch(c)  
        ax.set_axis_off()

    else:
        fig, ax = plt.subplots(1, 1, constrained_layout=True, figsize=(12, 4), sharex=True, sharey=True)

        ax.imshow(image, cmap='gray')
        y = dog_array[0]
        x = dog_array[1]
        r = dog_array[2]
        c = plt.Circle((x, y), r*5, color='red', linewidth=1, fill=False)
        ax.add_patch(c)  
        ax.set_axis_off()   
    
    plt.show()
    plt.close()
```

```python
log_txt, dog_txt, doh_txt = GetFileName(names, "Result")
thresh = [file[-10:-4] for file in log_txt[0]]

log_tracks = GetTracks(log_txt)
dog_tracks = GetTracks(dog_txt)
doh_tracks = GetTracks(doh_txt)
```

```python
missed_log = FindMissingBlobs(log_tracks, log_missing, names)
missed_dog = FindMissingBlobs(dog_tracks, dog_missing, names)
missed_doh = FindMissingBlobs(doh_tracks, doh_missing, names)

print("Log missed\n", missed_log, "\nDog missed\n", missed_dog, "\nDoh missed\n", missed_doh)
```

## Comment of the above result
The function FindMissingBlobs compares the true tracks(coordinates and radius),stored in the text files created along with the test images, with the tracks founded by the three algorithms; the comparison is based on the difference in distance between the centers of the blobs and the true radius, i.e, if the distance is shorter than the radius we can assume that the blob found by the algorithm is the same present in the file of the true tracks. The other action of the function is stored the information about the blobs missed by the algorithm, to be precise it tells us the image's number (in which there are missed blobs), the coordinates of these blobs and their gray levels; based on the results above, we can see that in general the algorithm missed some blobs because they are too close to others with bigger radius which hide the smaller blobs.
