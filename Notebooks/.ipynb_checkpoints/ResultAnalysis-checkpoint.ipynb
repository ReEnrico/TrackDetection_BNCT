```python
import os
import glob

import matplotlib.pyplot as plt
import numpy as np
from skimage import io


def upload_main_file(name_array: list):
    """Upload all files used in the program
       The directories where they are saved are set by default
       Three array are returned:
        the images as numpy array
        the texts files with blobs information (coordinates, radius and color)
        the number of blobs in each image
    """
    image_array = []
    text_array = []
    lenght_array = []
    for name in name_array:
        img = io.imread(os.path.join("..", "img_PseudoTracks", "test_{}.jpg".format(name)))
        image_array.append(img)
        txt = np.loadtxt(os.path.join("..", "img_PseudoTracks", "test_{}_coord.txt".format(name)))
        text_array.append(txt)
        lnght = len(txt)
        lenght_array.append(lnght)
        
    return image_array, text_array, lenght_array


def get_file_name(name_array: list, algorithm_name: str):
    """Find the names (string) of all text files contained in the repository
       selected throught the string 'algorithm_name'
    """
    algo_file = []
    for name in name_array:
        file = glob.glob(os.path.join("..", "Test", "Text", "First_Try", 
                                      "{algo_name}_Result".format(algo_name=algorithm_name), 
                                      "{name}_*.txt".format(name=name)))
        file.sort()
        algo_file.append(file)

    return algo_file

def get_thresh(file_array: list, name: str):
    """From each string (file name) get a substring which is
       the threshold used for analyzed the images    
    """
    thresh_list = []
    if name == "Doh":
        for file in file_array:
            image_thresh = []
            for txt in file:
                image_thresh.append(txt[-12:-4])
            thresh_list.append(image_thresh)
    else:
        for file in file_array:
            image_thresh = []
            for txt in file:
                image_thresh.append(txt[-10:-4])
            thresh_list.append(image_thresh)
        
    return thresh_list

def get_tracks(name_array: list):
    """Upload texts files with blob information created by the algorithms
       Using the string list created by the above function
    """
    track_array = []
    for i in range(len(name_array)):
        tmp = []
        for j in range(len(name_array[i])):
            dmp = np.loadtxt(name_array[i][j])
            tmp.append(dmp)
        track_array.append(tmp)
        
    return np.asarray(track_array, dtype=object)

def algorithm_information(
        name_array: list, track_array: np.array, 
        lenght_array: list, thresh_array: list):
    info_array = np.zeros((len(name_array), 4), dtype=object)
    for i in range(len(name_array)):
        difference = len(track_array[i][0]) - lenght_array[i]
        info_array[i] = np.array([name_array[i], float(thresh_array[i][0]), 
                                  lenght_array[i], difference], dtype=object)
    return info_array

def plot_histograms(
        first_array: np.array, 
        second_array: np.array, 
        third_array: np.array):
    fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, constrained_layout=True, figsize=(12,7))
    fig.suptitle("Distribution of threshold values")
    
    n, bins, patches = ax1.hist(first_array[:, 1])
    ax1.set_title("Log algorithm")
    ax1.set_xlabel("threshold")
    ax1.set_ylabel("how often")
    
    n, bins, patches = ax2.hist(second_array[:, 1])
    ax2.set_title("Doh algorithm")
    ax2.set_xlabel("threshold")
    ax2.set_ylabel("how often")
    
    n, bins, patches = ax3.hist(third_array[:, 1])
    ax3.set_title("Dog algorithm")
    ax3.set_xlabel("threshold")
    ax3.set_ylabel("how often")
    
    plt.show()
    plt.close()

def organizing_function(algo_info_array: np.array):
    extra_track = []
    right_track = []
    miss_track = []
    for i in range(len(algo_info_array)):
        if log_information[i][3] < 0:
            miss_track.append([int(algo_info_array[i][0]), algo_info_array[i][1]])
        elif log_information[i][3] == 0:
            right_track.append([int(algo_info_array[i][0]), algo_info_array[i][1]])
        else:
            extra_track.append([int(algo_info_array[i][0]), algo_info_array[i][1]])
            
    return extra_track, right_track, miss_track

def find_min_max_gray(blob_array: np.array, array: np.array):
    max_gray = np.zeros((len(blob_array), 2), dtype=object)
    #min_gray = np.zeros((len(blob_array), 2), dtype=object)
    for i in range(len(blob_array)):
        max_gray[i] = np.array([max(blob_array[i][:, 3]), array[i][1]], dtype=object)
        #min_gray[i] = min(blob_array[i][:, 3])
        
    return max_gray

names = ["{:03d}".format(i) for i in range(240)]
images, texts, lenghts = upload_main_file(names)

dog_txt = get_file_name(names, "Dog")
dog_thresh = get_thresh(dog_txt, "Dog")
dog_tracks = get_tracks(dog_txt)

doh_txt = get_file_name(names, "Doh")
doh_thresh = get_thresh(doh_txt, "Doh")
doh_tracks = get_tracks(doh_txt)

log_txt = get_file_name(names, "Log")
log_thresh = get_thresh(log_txt, "Log")
log_tracks = get_tracks(log_txt)


for i in range(len(log_tracks)):
    print(names[i], len(log_tracks[i]))

log_information = algorithm_information(names, log_tracks, lenghts, log_thresh)
doh_information = algorithm_information(names, doh_tracks, lenghts, doh_thresh)
dog_information = algorithm_information(names, dog_tracks, lenghts, dog_thresh)
print(log_information)

plot_histograms(log_information, doh_information, dog_information)

log_extra, log_correct, log_missed = organizing_function(log_information)
doh_extra, doh_correct, doh_missed = organizing_function(doh_information)
dog_extra, dog_correct, dog_missed = organizing_function(dog_information)
print(log_extra)

# get information of files with extra blob
# plot file with extra blob
# only Log

log_extra_files = [texts[log_extra[i][0]] for i in range(len(log_extra))]
bright_blob = find_min_max_gray(log_extra_files, log_extra)
print(log_extra)
print(bright_blob)
```
