```python
def get_all_needed(folder_name: str, number:int):
    name_list = glob.glob(os.path.join(folder_name, "Group_{}".format(number), 
                        "Results", "*.txt"))
    name_list.sort()
    result_array = [np.loadtxt(name) for name in name_list]
    name_list = [name[-14:-11] for name in name_list]
    
    image_array = [io.imread(os.path.join("..", "..", "img_PseudoTracks",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "..", "img_PseudoTracks",
                            "{}.txt".format(name))) for name in name_list]

    thresh_list = [glob.glob(os.path.join(folder_name, "Group_{}".format(number),
                    "{}_*.txt".format(name))) for name in name_list]
    thresh_list.sort()
    float_thresh = []
    for image in thresh_list:
        image_thresh = []
        for thresh in image:
            image_thresh.append(float(thresh[-10:-4]))
        float_thresh.append(image_thresh)
    lenght_min = len(float_thresh[0])
    for thresh in float_thresh:
        if len(thresh) < lenght_min:
            lenght_min = len(thresh)
    thresh_array = float_thresh[0][:lenght_min]

    track_by_thresh_file = []
    for thresh in thresh_array:
        image_file = []
        for name in name_list:
            if folder_name == "First":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.4f}.txt".format(name=name, thresh=thresh)))
            elif folder_name == "Second":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.5f}.txt".format(name=name, thresh=thresh)))
            image_file.append(text)
        track_by_thresh_file.append(image_file)

    return name_list, image_array, track_array, result_array, thresh_array, track_by_thresh_file

def false_negative(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_negative = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_negative = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_negative = []
            if not np.isscalar(track_array[i][j][0]):
                a_blob_i = [i for i in range(len(track_array[i][j]))]
                if not np.isscalar(true_array[j][0]):
                    for blob in true_array[j]:
                        y, x, r, c = blob
                        check = False                
                        for a_i in a_blob_i:
                            y_a, x_a, r_a = track_array[i][j][a_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                break
                        if check:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                        else:
                            false_negative.append(blob)
                else:
                    y, x, r, c = true_array[j]
                    check = False                
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            break
                    if check:
                        true.append(track_array[i][j][a_i])
                        a_blob_i.remove(a_i)
                    else:
                        false_negative.append(blob)

            else:
                if not np.isscalar(true_array[j][0]):
                    true_i = [i for i in range(len(true_array[j]))]
                    for t_i in true_i:
                        y, x, r, c = true_array[j][t_i]
                        y_a, x_a, r_a = track_array[i][j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j])
                            true_i.remove(t_i)
                            break
                    for t_i in true_i:
                        false_negative.append(true_array[j][t_i])
                else:
                    y, x, r, c = true_array[j]
                    y_a, x_a, r_a = track_array[i][j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_negative.append(true_array[j])

            array_true.append(true)
            array_false_negative.append(false_negative)
        total_true.append(array_true)
        total_false_negative.append(array_false_negative)
    return total_true, total_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_positive = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_positive = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_positive = []
            if not np.isscalar(track_array[i][j][0]):
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    for a_blob in track_array[i][j]:
                        y_a, x_a, r_a = a_blob
                        check = False
                        for b_i in blob_j:
                            y, x, r, c = true_array[j][b_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                true.append(a_blob)
                                blob_j.remove(b_i)
                                break
                        if not check:
                            false_positive.append(a_blob)
                else:
                    a_blob_i = [i for i in range(len(track_array[i][j]))]
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        y, x, r, c = true_array[j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                            break
                    for a_i in a_blob_i:
                        false_positive.append(track_array[i][j][a_i])                       
                
            else:
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    y_a, x_a, r_a = track_array[i][j]
                    check = False
                    for b_i in blob_j:
                        y, x, r, c = true_array[j][b_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            true.append(track_array[i][j])
                            blob_j.remove(b_i)
                            break
                    if not check:
                        false_positive.append(track_array[i][j])
                else:
                    y_a, x_a, r_a = track_array[i][j]
                    y, x, r, c = true_array[j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_positive.append(track_array[i][j])
                        
            array_true.append(true)
            array_false_positive.append(false_positive)
        total_true.append(array_true)
        total_false_positive.append(array_false_positive)
    return total_true, total_false_positive
```

```python
def get_all_needed(folder_name: str, number:int):
    name_list = glob.glob(os.path.join(folder_name, "Group_{}".format(number), 
                        "Results", "*.txt"))
    name_list.sort()
    result_array = [np.loadtxt(name) for name in name_list]
    name_list = [name[-14:-11] for name in name_list]
    
    image_array = [io.imread(os.path.join("..", "..", "img_PseudoTracks",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "..", "img_PseudoTracks",
                            "{}.txt".format(name))) for name in name_list]

    thresh_list = [glob.glob(os.path.join(folder_name, "Group_{}".format(number),
                    "{}_*.txt".format(name))) for name in name_list]
    thresh_list.sort()
    float_thresh = []
    for image in thresh_list:
        image_thresh = []
        for thresh in image:
            image_thresh.append(float(thresh[-10:-4]))
        float_thresh.append(image_thresh)
    lenght_min = len(float_thresh[0])
    for thresh in float_thresh:
        if len(thresh) < lenght_min:
            lenght_min = len(thresh)
    thresh_array = float_thresh[0][:lenght_min]

    track_by_thresh_file = []
    for thresh in thresh_array:
        image_file = []
        for name in name_list:
            if folder_name == "First":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.4f}.txt".format(name=name, thresh=thresh)))
            elif folder_name == "Second":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.5f}.txt".format(name=name, thresh=thresh)))
            image_file.append(text)
        track_by_thresh_file.append(image_file)

    return name_list, image_array, track_array, result_array, thresh_array, track_by_thresh_file

def false_negative(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_negative = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_negative = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_negative = []
            if not np.isscalar(track_array[i][j][0]):
                a_blob_i = [i for i in range(len(track_array[i][j]))]
                if not np.isscalar(true_array[j][0]):
                    for blob in true_array[j]:
                        y, x, r, c = blob
                        check = False                
                        for a_i in a_blob_i:
                            y_a, x_a, r_a = track_array[i][j][a_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                break
                        if check:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                        else:
                            false_negative.append(blob)
                else:
                    y, x, r, c = true_array[j]
                    check = False                
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            break
                    if check:
                        true.append(track_array[i][j][a_i])
                        a_blob_i.remove(a_i)
                    else:
                        false_negative.append(blob)

            else:
                if not np.isscalar(true_array[j][0]):
                    true_i = [i for i in range(len(true_array[j]))]
                    for t_i in true_i:
                        y, x, r, c = true_array[j][t_i]
                        y_a, x_a, r_a = track_array[i][j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j])
                            true_i.remove(t_i)
                            break
                    for t_i in true_i:
                        false_negative.append(true_array[j][t_i])
                else:
                    y, x, r, c = true_array[j]
                    y_a, x_a, r_a = track_array[i][j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_negative.append(true_array[j])

            array_true.append(true)
            array_false_negative.append(false_negative)
        total_true.append(array_true)
        total_false_negative.append(array_false_negative)
    return total_true, total_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_positive = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_positive = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_positive = []
            if not np.isscalar(track_array[i][j][0]):
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    for a_blob in track_array[i][j]:
                        y_a, x_a, r_a = a_blob
                        check = False
                        for b_i in blob_j:
                            y, x, r, c = true_array[j][b_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                true.append(a_blob)
                                blob_j.remove(b_i)
                                break
                        if not check:
                            false_positive.append(a_blob)
                else:
                    a_blob_i = [i for i in range(len(track_array[i][j]))]
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        y, x, r, c = true_array[j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                            break
                    for a_i in a_blob_i:
                        false_positive.append(track_array[i][j][a_i])                       
                
            else:
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    y_a, x_a, r_a = track_array[i][j]
                    check = False
                    for b_i in blob_j:
                        y, x, r, c = true_array[j][b_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            true.append(track_array[i][j])
                            blob_j.remove(b_i)
                            break
                    if not check:
                        false_positive.append(track_array[i][j])
                else:
                    y_a, x_a, r_a = track_array[i][j]
                    y, x, r, c = true_array[j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_positive.append(track_array[i][j])
                        
            array_true.append(true)
            array_false_positive.append(false_positive)
        total_true.append(array_true)
        total_false_positive.append(array_false_positive)
    return total_true, total_false_positive

def global_plot(result_array: np.array,
                true_track_array: np.array,
                false_negative_array: np.array, 
                false_positive_array: np.array, 
                thresh_list: list,
                number: int,
                folder_name: str):
    # analisi globale usando la media dei risultati
    
    blob_array = np.zeros((len(thresh_list), len(result_array)))
    for i in range(len(result_array)):
        for j in range(1, len(thresh_list)+1):          
                blob_array[j-1][i] = result_array[i][-j][1]
    mean_blob = [int(blob.mean()) for blob in blob_array]
    
    time = np.zeros((len(thresh_list), len(result_array)))
    for i in range(len(result_array)):
        for j in range(1, len(thresh_list)+1):          
                time[j-1][i] = result_array[i][-j][2]
    mean_time = [int(blob.mean()) for blob in time]
    
    true_mean_thresh = []
    for i in range(len(true_track_array)):
        mean_image = []
        for j in range(len(true_track_array[i])):
            mean_image.append(len(true_track_array[i][j]))
        mean = int(np.mean(mean_image))
        true_mean_thresh.append(mean)

    negative_mean_thresh = []
    for i in range(len(false_negative_array)):
        mean_image = []
        for j in range(len(false_negative_array[i])):
            mean_image.append(len(false_negative_array[i][j]))
        mean = int(np.mean(mean_image))
        negative_mean_thresh.append(mean)

    positive_mean_thresh = []
    for i in range(len(false_positive_array)):
        mean_image = []
        for j in range(len(false_positive_array[i])):
            mean_image.append(len(false_positive_array[i][j]))
        mean = int(np.mean(mean_image))
        positive_mean_thresh.append(mean)
    
    thresh_string = ["{:.5f}".format(thresh) for thresh in thresh_list]
    np_thresh = np.asarray(thresh_list)
    bottom_negative = true_mean_thresh
    bottom_positive = [num1+num2 for num1, num2 in zip(true_mean_thresh, negative_mean_thresh)]
    # all data needed     
    
    fig, (ax1, ax2) = plt.subplots(2, 1, constrained_layout=True, figsize=(7, 6))
    fig.suptitle("Algorithm's beheviour", y=1.05)

    ax1.plot(thresh_list, mean_blob, color='red', label="Tracks number")    
    ax1.set_title("Tracks number and process time in function of threshold parameter")
    ax1.set_xlim(left=np.min(thresh_list), right=np.max(thresh_list))
    if folder_name=="First":
        ax1.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.0002))
    elif folder_name=="Second":
        ax1.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.00002))
    ax1.tick_params(axis='x', labelrotation=65)
    ax1.set_xlabel("Threshold")
    ax1.set_ylabel("Tracks number")
    ax1.set_yscale("log")
    ax1.legend(loc=1)
    
    ax3 = ax1.twinx()
    ax3.plot(thresh_list, mean_time, color='blue', label="Execution time")
    ax3.set_ylabel("Time(s)")
    ax3.legend(loc=7)
    
    p1 = ax2.bar(thresh_string, true_mean_thresh, color="green", label="True")
    p2 = ax2.bar(thresh_string, negative_mean_thresh, bottom=bottom_negative, 
                 color="blue", label="False Negative")
    p3 = ax2.bar(thresh_string, positive_mean_thresh, bottom=bottom_positive,
                 color="red", label="False Positive")  
    ax2.set_xlim(left=np.min(thresh_list), right=np.max(thresh_list))
    ax2.set_xticks(thresh_string)
    ax2.set_xticklabels(thresh_string, rotation=65)
    ax2.set_xlabel("Threshold Value")
    ax2.set_ylabel("Number of Tracks")
    ax2.set_title("True, false segative, false positive tracks ratio")
    #ax2.set_xscale("log")
    ax2.set_yscale("log")
    ax2.legend()
    #plt.autoscale(enable=True, axis='x', tight=None)
    #plt.savefig("result_histogram_{}".format(number), bbox_inches='tight')
    plt.show()
    plt.close()
    
def number_track_plot(text_track_array: np.array,
                      algo_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list,
                      number: int,
                      folder_name: str):

    error_true_neg_thresh = []
    error_true_pos_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        error_true_pos_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            total_tracks = len(algo_track_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            error_true_neg = np.round((true_found/len(text_track_array[j]))*100, 2)
            error_true_pos = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/len(text_track_array[j]))*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
            
            error_true_neg_image.append(error_true_neg)
            error_true_pos_image.append(error_true_pos)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        mean_true_pos = np.round(np.mean(error_true_pos_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        
        error_true_neg_thresh.append(mean_true_neg)
        error_true_pos_thresh.append(mean_true_pos)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    percent_function1 = np.zeros((len(thresh_list)))
    percent_function2 = np.zeros((len(thresh_list)))
    for i in range(len(thresh_list)):
        percent_function1[i] = (error_true_neg_thresh[i]*(100-error_negative_thresh[i])*(100-error_positive_thresh[i])/1000000)*100
        percent_function2[i] = (error_true_pos_thresh[i]*(100-error_negative_thresh[i])*(100-error_positive_thresh[i])/1000000)*100
    
    fig, ax = plt.subplots(1, 1, figsize=(7, 3))
    fig.suptitle("True, false negative and false positive tracks percentage", y=1.05)

    ax.plot(thresh_list, error_true_neg_thresh, color="blue", label="True tracks")
    ax.plot(thresh_list, error_negative_thresh, color="green", label="False negative")
    ax.plot(thresh_list, error_positive_thresh, color="red", label="False positive")
    ax.set_xlim(left=np.min(thresh_list), right=np.max(thresh_list))
    if folder_name=="First":
        ax.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.0002))
    elif folder_name=="Second":
        ax.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.00002))
    ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Percentage")
    ax.legend() 
    plt.tight_layout()
    plt.savefig("all_track_{}".format(number), bbox_inches='tight')
    #plt.show()
    plt.close()
    
    fig, ax = plt.subplots(1, 1, figsize=(7, 3))
    fig.suptitle("Global percentage function", y=1.05)
    ax.plot(thresh_list, percent_function1, color="green", label="From true tracks number")
    ax.plot(thresh_list, percent_function2, color="red", label="From algorithm's tracks number")
    ax.set_xlim(left=np.min(thresh_list), right=np.max(thresh_list))
    if folder_name=="First":
        ax.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.0002))
    elif folder_name=="Second":
        ax.set_xticks(np.arange(np.min(thresh_list), np.max(thresh_list), 0.00002))
    ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Percentage")
    plt.tight_layout()
    plt.legend()
    #plt.savefig("all_percent_{}".format(number), bbox_inches='tight')
    plt.show()
    plt.close()


count = 1
for i in range(1, 6):
    print("Group {}".format(count))
    names, images, texts, result_array, thresh, track_array = get_all_needed("First", i)
    true_from_fals_neg, false_negative_track = false_negative(track_array, texts)
    true_from_fals_pos, false_positive_track = false_positive(track_array, texts)

    global_plot(result_array,
                true_from_fals_neg, 
                false_negative_track,
                false_positive_track,
                thresh,
                i,
                "First")
    number_track_plot(texts,
                      track_array,
                      true_from_fals_neg,
                      false_negative_track,
                      false_positive_track,
                      thresh,
                      i,
                      "First")
    count += 1
    names = None 
    images = None
    texts = None
    result_array = None
    thresh = None
    track_array = None
    true_from_fals_neg = None
    false_negative_track = None
    true_from_fals_pos = None
    false_positive_track = None
    time.sleep(2)

```

```python
def get_all_needed1(folder_name: str, number:int):
    name_list = glob.glob(os.path.join(folder_name, "Group_{}".format(number), 
                        "Results", "*.txt"))
    name_list.sort()
    name_list = [name[-14:-11] for name in name_list]

    image_array = [io.imread(os.path.join("..", "..", "img_PseudoTracks",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "..", "img_PseudoTracks",
                            "{}.txt".format(name))) for name in name_list]

    thresh_list = [glob.glob(os.path.join(folder_name, "Group_{}".format(number),
                    "{}_*.txt".format(name))) for name in name_list]
    thresh_list.sort()
    float_thresh = []
    for image in thresh_list:
        image_thresh = []
        for thresh in image:
            image_thresh.append(float(thresh[-10:-4]))
        float_thresh.append(image_thresh)
    lenght_min = len(float_thresh[0])
    for thresh in float_thresh:
        if len(thresh) < lenght_min:
            lenght_min = len(thresh)
    thresh_array = float_thresh[0][:lenght_min][:4]
    track_by_thresh_file = []
    for thresh in thresh_array:
        image_file = []
        for name in name_list:
            if folder_name == "First":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.4f}.txt".format(name=name, thresh=thresh)))
            elif folder_name == "Second":
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.5f}.txt".format(name=name, thresh=thresh)))
            image_file.append(text)
        track_by_thresh_file.append(image_file)

    return name_list, image_array, track_array, thresh_array, track_by_thresh_file

def show_blob(text: np.array,
              track_array: list,
              false_negative: list,
              false_positive: list,
              image: np.array,
              thresh_list: list,
              name: str,
              index):
    index = index
    for i in range(len(thresh_list)):
        fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(10, 7))
        fig.suptitle("Image {}, true tracks {}, threshold {:.5f}".format(name,
                                                                         len(text),
                                                                         thresh_list[i]), y=0.9)
        ax1.imshow(image, cmap="gray")
        ax1.scatter(text[:, 1], text[:, 0], s=text[:, 2], color="green")                  
        for blob in track_array[i][index]:
            y, x, r = blob
            c = plt.Circle((x, y), r, color="orange", fill=False)
            ax1.add_patch(c)
        ax1.set_title("Tracks found {}".format(len(track_array[i][index])))
        
        ax2.imshow(image, cmap="gray")
        if len(np.array(false_positive[i][index]))!=0:
            ax2.scatter(np.array(false_negative[i][index])[:, 1],
                        np.array(false_negative[i][index])[:, 0],
                        s=np.array(false_negative[i][index])[:, 2], color="blue")
            if not np.isscalar(np.array(false_positive[i][index])[0]):
                ax2.scatter(np.array(false_positive[i][index])[:, 1],
                                     np.array(false_positive[i][index])[:, 0],
                                     s=np.array(false_positive[i][index])[:, 2], color="red")
            else:
                ax2.scatter(np.array(false_positive[i][index])[1],
                            np.array(false_positive[i][index])[0],
                            s=np.array(false_positive[i][index])[2], color="red")
            ax2.set_title("False Negative {}, False Positive {}".format(len(false_negative[i][index]),
                                                                       len(false_positive[i][index])))
        else:
            ax2.scatter(np.array(false_negative[i][index])[:, 1],
                        np.array(false_negative[i][index])[:, 0],
                        s=np.array(false_negative[i][index])[:, 2], color="blue")
            ax2.set_title("False Negative {}, False Positive 0".format(len(false_negative[i][index]),
                                                                       len(false_positive[i][index])))
        plt.show()
        plt.close()
```
