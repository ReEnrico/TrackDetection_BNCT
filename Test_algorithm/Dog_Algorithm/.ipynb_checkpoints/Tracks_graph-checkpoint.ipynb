```python
import os
import glob
import time
import sys
sys.path.insert(0, "..")

import matplotlib.pyplot as plt
import numpy as np
from skimage import io
```

```python
def get_all_needed(folder_name: str, number:int):
    name_list = glob.glob(os.path.join(folder_name, "Group_{}".format(number), 
                        "Results", "*.txt"))
    name_list.sort()
    
    name_list = [name[-14:-11] for name in name_list[:10]]
    image_array = [io.imread(os.path.join("..", "..", "img_PseudoTracks",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "..", "img_PseudoTracks",
                            "{}.txt".format(name))) for name in name_list]
    
    thresh_list = [glob.glob(os.path.join(folder_name, "Group_{}".format(number),
                    "{}_*.txt".format(name))) for name in name_list]
    thresh_list.sort()
    float_thresh = []
    for thresh in thresh_list[0][:21]:
        float_thresh.append(float(thresh[-9:-4]))
    
    track_by_thresh_file = []
    for thresh in float_thresh:
        image_file = []
        for name in name_list:
            if os.path.isfile(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.3f}.txt".format(name=name, thresh=thresh))):
                text = np.loadtxt(os.path.join(folder_name, "Group_{}".format(number),
                            "{name}_{thresh:.3f}.txt".format(name=name, thresh=thresh)))
                image_file.append(text)
        track_by_thresh_file.append(image_file)
    
    return name_list, image_array, track_array, float_thresh, track_by_thresh_file

def false_negative(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_negative = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_negative = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_negative = []
            if not np.isscalar(track_array[i][j][0]):
                a_blob_i = [i for i in range(len(track_array[i][j]))]
                if not np.isscalar(true_array[j][0]):
                    for blob in true_array[j]:
                        y, x, r, c = blob
                        check = False                
                        for a_i in a_blob_i:
                            y_a, x_a, r_a = track_array[i][j][a_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                break
                        if check:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                        else:
                            false_negative.append(blob)
                else:
                    y, x, r, c = true_array[j]
                    check = False                
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            break
                    if check:
                        true.append(track_array[i][j][a_i])
                        a_blob_i.remove(a_i)
                    else:
                        false_negative.append(blob)

            else:
                if not np.isscalar(true_array[j][0]):
                    true_i = [i for i in range(len(true_array[j]))]
                    for t_i in true_i:
                        y, x, r, c = true_array[j][t_i]
                        y_a, x_a, r_a = track_array[i][j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j])
                            true_i.remove(t_i)
                            break
                    for t_i in true_i:
                        false_negative.append(true_array[j][t_i])
                else:
                    y, x, r, c = true_array[j]
                    y_a, x_a, r_a = track_array[i][j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_negative.append(true_array[j])

            array_true.append(true)
            array_false_negative.append(false_negative)
        total_true.append(array_true)
        total_false_negative.append(array_false_negative)
    return total_true, total_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_positive = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_positive = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_positive = []
            if not np.isscalar(track_array[i][j][0]):
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    for a_blob in track_array[i][j]:
                        y_a, x_a, r_a = a_blob
                        check = False
                        for b_i in blob_j:
                            y, x, r, c = true_array[j][b_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                true.append(a_blob)
                                blob_j.remove(b_i)
                                break
                        if not check:
                            false_positive.append(a_blob)
                else:
                    a_blob_i = [i for i in range(len(track_array[i][j]))]
                    for a_i in a_blob_i:
                        y_a, x_a, r_a = track_array[i][j][a_i]
                        y, x, r, c = true_array[j]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i][j][a_i])
                            a_blob_i.remove(a_i)
                            break
                    for a_i in a_blob_i:
                        false_positive.append(track_array[i][j][a_i])                       
                
            else:
                if not np.isscalar(true_array[j][0]):
                    blob_j = [i for i in range(len(true_array[j]))]
                    y_a, x_a, r_a = track_array[i][j]
                    check = False
                    for b_i in blob_j:
                        y, x, r, c = true_array[j][b_i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            check = True
                            true.append(track_array[i][j])
                            blob_j.remove(b_i)
                            break
                    if not check:
                        false_positive.append(track_array[i][j])
                else:
                    y_a, x_a, r_a = track_array[i][j]
                    y, x, r, c = true_array[j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    if d2 <= (r)**2:
                        true.append(track_array[i][j])
                    else:
                        false_positive.append(track_array[i][j])
                        
            array_true.append(true)
            array_false_positive.append(false_positive)
        total_true.append(array_true)
        total_false_positive.append(array_false_positive)
    return total_true, total_false_positive


def using_total_track(algo_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list,
                      number: int):

    controll = []
    error_true_pos_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_positive_array)):
        error_true_pos_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_positive_array[i])):
            total_tracks = len(algo_track_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            error_true_pos = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/total_tracks)*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
            
            error_true_pos_image.append(error_true_pos)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_pos = np.round(np.mean(error_true_pos_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total = mean_true_pos + mean_negative + mean_positive
        
        controll.append(total)
        error_true_pos_thresh.append(mean_true_pos)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    fig.suptitle("Percentuali rapportate al numero di tracce trovate", y=1.05)

    ax.plot(thresh_list, error_true_pos_thresh, color="blue", label="True tracks")
    ax.plot(thresh_list, error_negative_thresh, color="green", label="False negative")
    ax.plot(thresh_list, error_positive_thresh, color="red", label="False positive")
    ax.plot(thresh_list, controll, color="black", label="Controll percentage")
    ax.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Percentage")
    ax.set_yscale("log")
    ax.legend() 
    plt.tight_layout()
    plt.savefig("usando_tracce_trovate".format(number), bbox_inches='tight')
    plt.show()
    plt.close()    
    
def using_true_track(text_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list,
                      number: int):

    controll = []
    error_true_neg_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            error_true_neg = np.round((true_found/len(text_track_array[j]))*100, 2)
            error_negative = np.round((false_negative/len(text_track_array[j]))*100, 2)
            error_positive = np.round((false_positive/len(text_track_array[j]))*100, 2)
            
            error_true_neg_image.append(error_true_neg)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total = mean_true_neg + mean_negative + mean_positive
        
        controll.append(total)
        error_true_neg_thresh.append(mean_true_neg)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    fig.suptitle("Percentuali rapportate al numero di tracce vere", y=1.05)

    ax.plot(thresh_list, error_true_neg_thresh, color="blue", label="True tracks")
    ax.plot(thresh_list, error_negative_thresh, color="green", label="False negative")
    ax.plot(thresh_list, error_positive_thresh, color="red", label="False positive")
    ax.plot(thresh_list, controll, color="black", label="Controll percentage")
    ax.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Percentage")
    ax.set_yscale("log")
    ax.legend() 
    plt.tight_layout()
    plt.savefig("usando_numero_vero".format(number), bbox_inches='tight')
    plt.show()
    plt.close()    
    
    
def using_both_track(text_track_array: np.array,
                      algo_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list,
                      number: int):

    controll_ng = []
    controll_ps = []
    error_true_neg_thresh = []
    error_true_pos_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        error_true_pos_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            total_tracks = len(algo_track_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            error_true_neg = np.round((true_found/len(text_track_array[j]))*100, 2)
            error_true_pos = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/len(text_track_array[j]))*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
            
            error_true_neg_image.append(error_true_neg)
            error_true_pos_image.append(error_true_pos)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        mean_true_pos = np.round(np.mean(error_true_pos_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total_ng = mean_true_neg + mean_negative
        total_ps = mean_true_pos + mean_positive
        
        controll_ng.append(total_ng)
        controll_ps.append(total_ps)
        error_true_neg_thresh.append(mean_true_neg)
        error_true_pos_thresh.append(mean_true_pos)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
    fig.suptitle("Percentuali separate", y=1.05)

    ax1.plot(thresh_list, error_true_neg_thresh, color="blue", label="True tracks")
    ax1.plot(thresh_list, error_negative_thresh, color="green", label="False negative")
    ax1.plot(thresh_list, controll_ng, color="black", label="controll percent")
    ax1.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax1.tick_params(axis='x', labelrotation=65)
    ax1.set_xlabel("Threshold")
    ax1.set_ylabel("Percentage")
    ax1.set_title("Tracce giuste e falsi negativi rapportate al numero vero di tracce esistenti")
    ax1.legend() 

    ax2.plot(thresh_list, error_true_pos_thresh, color="blue", label="True tracks")
    ax2.plot(thresh_list, error_positive_thresh, color="red", label="False positive")
    ax2.plot(thresh_list, controll_ps, color="black", label="controll percent")
    ax2.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax2.tick_params(axis='x', labelrotation=65)
    ax2.set_xlabel("Threshold")
    ax2.set_ylabel("Percentage")
    ax2.set_title("Tracce giuste e falsi negativi rapportate al numero vero di tracce trovate")
    ax2.legend() 
    
    plt.tight_layout()
    plt.savefig("percentuali_separati".format(number), bbox_inches='tight')
    plt.show()
    plt.close()

def alternate_track(text_track_array: np.array,
                      algo_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list,
                      number: int):
    
    controll = []
    error_true_neg_thresh = []
    #error_true_pos_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        #error_true_pos_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            total_tracks = len(algo_track_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            if len(text_track_array[j]) >= total_tracks:
                error_true_neg = np.round((true_found/len(text_track_array[j]))*100, 2)
                #error_true_pos = np.round((true_found/total_tracks)*100, 2)
                error_negative = np.round((false_negative/len(text_track_array[j]))*100, 2)
                error_positive = np.round((false_positive/len(text_track_array[j]))*100, 2)
            else:
                error_true_neg = np.round((true_found/total_tracks)*100, 2)
                #error_true_pos = np.round((true_found/total_tracks)*100, 2)
                error_negative = np.round((false_negative/total_tracks)*100, 2)
                error_positive = np.round((false_positive/total_tracks)*100, 2)
            
            error_true_neg_image.append(error_true_neg)
            #error_true_pos_image.append(error_true_pos)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        #mean_true_pos = np.round(np.mean(error_true_pos_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total = mean_true_neg + mean_negative + mean_positive
        
        controll.append(total)
        error_true_neg_thresh.append(mean_true_neg)
        #error_true_pos_thresh.append(mean_true_pos)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    
    fig, ax1 = plt.subplots(1, 1, figsize=(10, 8))
    fig.suptitle("Percentuali calcolate usando prima il numero vero, poi le tracce trovate", y=1.05)

    ax1.plot(thresh_list, error_true_neg_thresh, color="blue", label="True tracks")
    ax1.plot(thresh_list, error_negative_thresh, color="green", label="False negative")
    ax1.plot(thresh_list, error_positive_thresh, color="red", label="False positive")
    ax1.plot(thresh_list, controll, color="black", label="Controll percentage")
    ax1.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax1.tick_params(axis='x', labelrotation=65)
    ax1.set_xlabel("Threshold")
    ax1.set_ylabel("Percentage")
    ax1.set_title("Tracce giuste, falsi negativi e falsi positivi")
    ax1.legend() 
    
    plt.tight_layout()
    plt.savefig("percentuali_miste".format(number), bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
names, images, texts, thresh, track_array = get_all_needed("First", 1)
true_from_fals_neg, false_negative_track = false_negative(track_array, texts)
true_from_fals_pos, false_positive_track = false_positive(track_array, texts)
```
