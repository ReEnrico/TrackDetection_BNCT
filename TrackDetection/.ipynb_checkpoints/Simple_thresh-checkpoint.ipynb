```python
import os
import glob
import time
import sys
sys.path.insert(0, "..")

import matplotlib.pyplot as plt
import numpy as np
from skimage import io
```

```python
def get_all_needed(folder_name: str):
    
    name_list = ["{:03d}".format(i) for i in range(240)]
    image_array = [io.imread(os.path.join("..", "img",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "img",
                            "{}.txt".format(name))) for name in name_list]
    
    return name_list, image_array, track_array

def get_thresh_file(folder_name: str, true_track: list):

    file_name = glob.glob(os.path.join(folder_name, "*.txt")) 
    file_name = [file[len(folder_name)+1:len(folder_name)+4] for file in file_name]
    file_name = list(dict.fromkeys(file_name))
    
    track_file = []
    thresh_list = []
    for i in range(len(true_track)):
        txt_name = glob.glob(os.path.join(folder_name, "{}*.txt".format(file_name[i])))
        txt_name.sort()
        text = [np.loadtxt(file) for file in txt_name]
        thresh = [file[-10:-4] for file in txt_name[:len(text)]]
        track_file.append(text)
        thresh_list.append(thresh)
        
    return track_file, thresh_list

def false_negative(track_array: np.array, 
                   true_array: np.array):
    
    array_true = []
    array_false_negative = []
    # loop over the images
    for i in range(len(track_array)):
        true = []
        false_negative = []
        if len(true_array[i]) != 0:
            if not np.isscalar(true_array[i][0]):
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto e con più di un blob
                        algo_index = [k for k in range(len(track_array[i]))]
                        for blob in true_array[i]:
                            y, x, r = blob
                            check = False
                            for a_i in algo_index:
                                y_a, x_a, r_a = track_array[i][a_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(track_array[i][a_i])
                                    algo_index.remove(a_i)
                                    break
                            if not check:
                                false_negative.append(blob)
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        true_index = [k for k in range(len(true_array[i]))]
                        for t_i in true_index:
                            y, x, r = true_array[i][t_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i])
                                true_index.remove(t_i)
                                break
                        for t_i in true_index:
                            false_negative.append(true_array[t_i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                else:
                    # immagine non vuota e con più di un blob
                    # file algoritmo vuoto
                    for blob in true_array[i]:
                        false_negative.append(blob)
                    array_true.append(true)
                    array_false_negative.append(false_negative)
                        
            else:
                y, x, r = true_array[i]
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto con più di un blob
                        check = False
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(a_blob)
                                check = True
                                break
                        if not check:
                            false_negative.append(true_array[i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto e con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_negative.append(true_array[i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)  
                            
                else:
                    # immagine non vuota e con un solo blob
                    # file algoritmo vuoto
                    false_negative.append(true_array[i])
                    array_false_negative.append(false_negative)
                        
        else:
            # immagine vuota
            # file algoritmo non importante
            array_true.append(true)
            array_false_negative.append(false_negative)
                
    return array_true, array_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    
    array_true = []
    array_false_positive = []
    # loop over the images
    for i in range(len(track_array)):
        true = []
        false_positive = []
        if len(true_array[i]) != 0:
            if not np.isscalar(true_array[i][0]):
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto e con più di un blob
                        true_index = [k for k in range(len(true_array[i]))]
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            check = False
                            for t_i in true_index:
                                y, x, r = true_array[i][t_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(a_blob)
                                    true_index.remove(t_i)
                                    break
                            if not check:
                                false_positive.append(a_blob)
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        check = False
                        for blob in true_array[i]:
                            y, x, r = blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i])
                                check = True
                                break
                        if not check:
                            false_positive.append(track_array[i])
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                            
                else:
                    # immagine non vuota e con più di un blob
                    # file algoritmo vuoto
                    array_true.append(true)
                    array_false_positive.append(false_positive)
                        
            else:
                y, x, r = true_array[i]
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuota e con più di un blob
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(a_blob)
                            else:
                                false_positive.append(a_blob)
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                           
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto e con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_positive.append(track_array[i])
                        array_true.append(true)
                        array_false_positive.append(false_positive)  
                            
                else:
                    # immagine non vuota e con un solo blob
                    # file algoritmo vuoto
                    array_true.append(true)
                    array_false_positive.append(false_positive)
                        
        else:
            array_true.append(true)
            if len(track_array[i]) != 0:
                if not np.isscalar(track_array[i][0]):
                    # immagine vuota
                    # file algortimo non vuoto e con più di un blob
                    for a_blob in track_array[i]:
                        false_positive.append(a_blob)
                else:
                    # immagine vuota
                    # file algoritmo non vuoto e con un blob
                    false_positive.append(track_array[i])
            array_false_positive.append(false_positive)
        
    return array_true, array_false_positive

def number_track_plot(text_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      name_list: list,
                      algo_name: str):

    controll = []
    true_track = []
    false_negative_track = []
    false_positive_track = []
    for i in range(len(name_list)):
        false_negative = len(false_negative_array[i])
        false_positive = len(false_positive_array[i])
        true_found = len(true_track_array[i])
        total_tracks = len(text_track_array[i]) + false_positive
        
        if total_tracks != 0:
            error_true_neg = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/total_tracks)*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
            total = error_true_neg + error_negative + error_positive
        
        else:
            error_true_neg = 100
            error_negative = false_negative
            error_positive = false_positive
            total = total = error_true_neg + error_negative + error_positive            
            
        controll.append(total)
        true_track.append(error_true_neg)
        false_negative_track.append(error_negative)
        false_positive_track.append(error_positive)
    
    mean_true = np.mean(np.asarray(true_track))
    std_true = np.std(np.asarray(true_track))
    mean_positive = np.mean(np.asarray(false_positive_track))
    std_positive = np.std(np.asarray(false_positive_track))
    mean_negative = np.mean(np.asarray(false_negative_track))
    std_negative = np.std(np.asarray(false_negative_track))
    name = [int(name) for name in name_list]

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))
    fig.suptitle("{} histogram".format(algo_name), y=1.05)

    ax1.hist(true_track, bins=15, color='b', edgecolor='black', label="True tracks")
    ax1.axvline(np.median(true_track), color="black", linestyle="--", alpha=0.25)
    ax1.set_xlim((0, 100))
    ax1.set_xlabel("Percentage")
    ax1.set_ylabel("Number of images")
    ax1.set_title("True tracks mean {:.2f} +- {:.2f}".format(np.median(true_track),
                                                             np.std(true_track)), y=1.05)
    ax1.legend()
    
    ax2.hist(false_negative_track, bins=15, color='g', edgecolor='black', label="False negative")
    ax2.axvline(np.median(false_negative_track), color="black", linestyle="--", alpha=0.25)
    ax2.set_xlim((0, 100))
    ax2.set_xlabel("Percentage")
    ax2.set_ylabel("Number of images")
    ax2.set_title("False negative tracks mean {:.2f} +- {:.2f}".format(np.median(false_negative_track),
                                                                       np.std(false_negative_track)), y=1.05)
    ax2.legend()
    
    ax3.hist(false_positive_track, bins=15, color='r', edgecolor='black', label="False positive")
    ax3.axvline(np.median(false_positive_track), color="black", linestyle="--", alpha=0.25)
    ax3.set_xlim((0, 100))
    ax3.set_xlabel("Percentage")
    ax3.set_ylabel("Number of images")
    ax3.set_title("False positive {:.2f} +- {:.2f}".format(np.median(false_positive_track),
                                                           np.std(false_positive_track)), y=1.05)
    ax3.legend()
    
    plt.tight_layout()
    #plt.savefig("Wrong_tre_hist_{}".format(algo_name), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return true_track, false_negative_track, false_positive_track

def plot_image(text: np.array,
               algo_track: np.array,
               algo_false_negative_track: np.array,
               algo_false_positive_track: np.array,               
               image: np.array,
               name: str):
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 15))
    fig.suptitle("Image {}".format(name))
    
    ax1.imshow(image, cmap="gray")
    if len(text) != 0:
        ax1.scatter(text[:, 1],
                    text[:, 0],
                    s=text[:, 2],
                    color='y',
                    alpha=0.30,
                    label="True")
    if len(algo_track) != 0:
        if not np.isscalar(algo_track[0]):
            ax1.scatter(algo_track[:, 1],
                        algo_track[:, 0],
                        s=algo_track[:, 2],
                        color='c',
                        alpha=0.30,
                        label="Algorithm")
        else:
            ax1.scatter(algo_track[1],
                        algo_track[0],
                        s=algo_track[2],
                        color='c',
                        alpha=0.30,
                        label="Algorithm")
        ax1.legend()
    ax1.set_title("True tracks {} and\n algorith tracks {}".format(len(text),
                                                               len(algo_track)))


    ax2.imshow(image, cmap="gray")
    if len(algo_false_negative_track) != 0:
        if np.isscalar(algo_false_negative_track[0]):
             ax2.scatter(algo_false_negative_track[1],
                         algo_false_negative_track[0],
                         s=algo_false_negative_track[2],
                         color='b',
                         label="False Negative")
        else:
            ax2.scatter(np.array(algo_false_negative_track)[:, 1],
                        np.array(algo_false_negative_track)[:, 0],
                        s=np.array(algo_false_negative_track)[:, 2],
                        color='b',
                        label="False Negative")
           
    if len(algo_false_positive_track) != 0:
        if not np.isscalar(algo_false_positive_track[0]):
            for blob in np.array(algo_false_positive_track):
                y, x, r = blob
                c = plt.Circle((x, y), r, color='r', fill=False)
                ax2.add_patch(c)
        else:
            y, x, r = np.array(algo_false_positive_track)
            c = plt.Circle((x, y), r, color='r', fill=False)
            ax2.add_patch(c)
    ax2.set_title("False positive {p_num} and\n false negative {n_num}".format(p_num=len(algo_false_positive_track),
                                                                               n_num=len(algo_false_negative_track)))

    plt.show()
    plt.close()
```

## Riassunto
Qui il threshold ha lavorato senza punti di riferimento partendo dal valore medio di sfondo e abbassandolo scartando tutto ciò che sta sopra (ricorda che l'effettivo algoritmo lavora con l'immagine ribaltata); funziona correttamente nelle immagini vuote, ma presenta troppi falsi negativi, l'abbassamento del parametro evidentemente è stato troppo rapido 0.05 e la soglia imposta per la pressa dati (#blob=1.3#vero troppo bassa). Comunuqe da questa analisi ricavo un valore massimo di threshold pari a 159.0675 facendo la media di tutti i valori più alti, ciò quelli per cui ha iniziato a scrivere qualcosa. L'intento è ora di creare una scansione comune, prima non possibile, che aumenti tale valore di una decina di volte (voglio avere una parte di analisi veloce) aumentandolo verso 163 (dagli istogrammi ricavati sembra essere quella la soglia massima di colore per i blob), sperando che non compaia troppo sporco, che rimane di piccole dimensioni eppure c'è un controllo sulla grandezza dell'area; proprio per la presenza di tale controllo (sebbene non sappia riportare il valore inserito in pixel) si può pensare di alzare la soglia minima di tolleranza in caso questa misura presenti troppi falsi postivi. I risultati ottenuti saranno differenti a quelli qui sopra: sarà percentuale trovata dui vere/negative/positive in funzione del threshold, qui è semplicemente per ogni singola immagine che percentuale c'è. 

La prima analisi (cartella Sbagliata) è piuttosto brutta, probabilmente la colpa è di exposure che tolgo. Non tocco le dimensioni delle regioni da cercare

```python
def get_all_needed(folder_name: str):
    
    name_list = glob.glob(os.path.join(folder_name, "*.txt"))
    name_list.sort()    
    names = [name[len(folder_name)+1:len(folder_name)+4] for name in name_list]
    names = list(dict.fromkeys(names))
    image_array = [io.imread(os.path.join("..","img",
                    "{}.jpg".format(name))) for name in names]
    track_array = [np.loadtxt(os.path.join("..", "img",
                    "{}_half.txt".format(name))) for name in names]
    string_tresh = [string[-10:-4] for string in name_list]
    thresh_list = list(dict.fromkeys(string_tresh))
    thresh_float = [float(num) for num in thresh_list]
    track_by_thresh_file = []
    for thresh in thresh_float:
        image_file = []
        for name in names:
            if os.path.isfile(os.path.join(folder_name,
                            "{name}_{thresh:.2f}.txt".format(name=name, thresh=thresh))):
                text = np.loadtxt(os.path.join(folder_name,
                            "{name}_{thresh:.2f}.txt".format(name=name, thresh=thresh)))
                image_file.append(text)
        track_by_thresh_file.append(image_file)

    return names, image_array, track_array, thresh_float, track_by_thresh_file

def false_negative(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_negative = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_negative = []
        # loop over the images
        # 240 file true and algorithm
        # true only index j
        for j in range(len(track_array[i])):
            true = []
            false_negative = []
            # image with blob
            if len(true_array[j]) != 0:
                # more than one blob
                if not np.isscalar(true_array[j][0]):
                    # tracks found
                    if len(track_array[i][j]) != 0:
                        # more than one track found
                        if not np.isscalar(track_array[i][j][0]):
                            a_blob_i = [i for i in range(len(track_array[i][j]))]
                            for blob in true_array[j]:
                                y, x, r = blob
                                check = False                
                                for a_i in a_blob_i:
                                    y_a, x_a, r_a = track_array[i][j][a_i]
                                    d2 = (x - x_a)**2 + (y - y_a)**2
                                    if d2 <= (r/2)**2:
                                        check = True
                                        break
                                if check:
                                    true.append(track_array[i][j][a_i])
                                    a_blob_i.remove(a_i)
                                else:
                                    false_negative.append(blob)
                        # one track found
                        else:
                            y_a, x_a, r_a = track_array[i][j]
                            blob_i = [i for i in range(len(true_array[j]))]
                            for b_i in blob_i:
                                y, x, r = true_array[j][b_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r/2)**2:
                                    true.append(track_array[i][j])
                                    blob_i.remove(b_i)
                                    break
                            for b_i in blob_i:
                                false_negative.append(true_array[j][b_i])
                    #no tracks found
                    else:
                        for blob in true_array[j]:
                            false_negative.append(blob)
                # one blob
                else:
                    y, x, r = true_array[j]
                    # tracks found
                    if len(track_array[i][j]) != 0:
                        # more than one track found
                        if not np.isscalar(track_array[i][j][0]):
                            check = False
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(a_blob)
                                    break
                            if not check:
                                false_negative.append(true_array[j])
                        # one track found
                        else:
                            y_a, x_a, r_a = track_array[i][j]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i][j])
                            else:
                                false_negative.append(true_array[j])
                    #no tracks found
                    else:
                        false_negative.append(true_array[j])            
            array_true.append(true)
            array_false_negative.append(false_negative)
        total_true.append(array_true)
        total_false_negative.append(array_false_negative)
    return total_true, total_false_negative

def false_positive(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_positive = []
    # loop over thresh value
    for i in range(len(track_array)):
        array_true = []
        array_false_positive = []
        # loop over the images
        # 240 file true and algorithm
        # true only index j
        for j in range(len(track_array[i])):
            true = []
            false_positive = []
            # image with blob
            if len(true_array[j]) != 0:
                # more than one blob
                if not np.isscalar(true_array[j][0]):
                    # tracks found
                    if len(track_array[i][j]) != 0:
                        # more than one track found
                        if not np.isscalar(track_array[i][j][0]):
                            blob_i = [i for i in range(len(true_array[j]))]
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                check = False
                                for b_i in blob_i:
                                    y, x, r = true_array[j][b_i]
                                    d2 = (x - x_a)**2 + (y - y_a)**2
                                    if d2 <= (r)**2:
                                        check = True
                                        true.append(a_blob)
                                        blob_i.remove(b_i)
                                        break
                                if not check:
                                    false_positive.append(a_blob)
                        # one track found
                        else:
                            y_a, x_a, r_a = track_array[i][j]
                            check = False
                            for blob in true_array[j]:
                                y, x, r = blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(track_array[i][j])
                                    break
                            if not check:
                                false_positive.append(track_array[i][j])
                # one blob
                else:
                    y, x, r = true_array[j]
                    # tracks found
                    if len(track_array[i][j]) != 0:
                        # more than one track found
                        if not np.isscalar(track_array[i][j][0]):
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    true.append(a_blob)
                                else:
                                    false_positive.append(a_blob)
                        # one track found
                        else:
                            y_a, x_a, r_a = track_array[i][j]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i][j])
                            else:
                                false_positive.append(track_array[i][j])
            # image empty
            else:
                # tracks found
                if len(track_array[i][j]) != 0:
                    # more than one track found
                    if not np.isscalar(track_array[i][j][0]):
                        for a_blob in track_array[i][j]:
                            false_positive.append(a_blob)                                                      
                    # one track found
                    else:
                        false_positive.append(track_array[i][j])
            array_true.append(true)
            array_false_positive.append(false_positive)
        total_true.append(array_true)
        total_false_positive.append(array_false_positive)
    return total_true, total_false_positive

def global_plot(true_track_array: np.array,
                false_negative_array: np.array, 
                false_positive_array: np.array, 
                thresh_list: list):
    
    # analisi globale usando la media dei risultati    
    true_mean_thresh = []
    for i in range(len(true_track_array)):
        mean_image = []
        for j in range(len(true_track_array[i])):
            if len(true_track_array[i][j]):
                mean_image.append(len(true_track_array[i][j]))
        mean = int(np.mean(mean_image))
        true_mean_thresh.append(mean)
    
    negative_mean_thresh = []
    for i in range(len(false_negative_array)):
        mean_image = []
        for j in range(len(false_negative_array[i])):
            mean_image.append(len(false_negative_array[i][j]))
        mean = int(np.mean(mean_image))
        negative_mean_thresh.append(mean)

    positive_mean_thresh = []
    for i in range(len(false_positive_array)):
        mean_image = []
        for j in range(len(false_positive_array[i])):
            mean_image.append(len(false_positive_array[i][j]))
        mean = int(np.mean(mean_image))
        positive_mean_thresh.append(mean)
    
    thresh_string = ["{:.2f}".format(thresh) for thresh in thresh_list]
    np_thresh = np.asarray(thresh_list)
    bottom_negative = true_mean_thresh
    bottom_positive = [num1+num2 for num1, num2 in zip(true_mean_thresh, negative_mean_thresh)]
    # all data needed     
    
    fig, ax = plt.subplots(1, 1, constrained_layout=True, figsize=(10, 7))
    fig.suptitle("Algorithm's beheviour", y=1.05)
   
    p1 = ax.bar(thresh_string, true_mean_thresh, color="green", label="True")
    p2 = ax.bar(thresh_string, negative_mean_thresh, bottom=bottom_negative, 
                 color="blue", label="False Negative")
    p3 = ax.bar(thresh_string, positive_mean_thresh, bottom=bottom_positive,
                 color="red", label="False Positive")  
    #ax.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    #ax.set_xticks(thresh_string)
    ax.set_xticklabels(thresh_string, rotation=65)
    ax.set_xlabel("Threshold Value")
    ax.set_ylabel("Number of Tracks")
    ax.set_title("True, false segative, false positive tracks ratio")
    #ax.set_xscale("log")
    ax.set_yscale("log")
    ax.legend()
    #plt.autoscale(enable=True, axis='x', tight=None)
    #plt.savefig("result_histogram_{}".format(number), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return None

def number_track_plot(text_track_array: np.array,
                      algo_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      thresh_list: list):

    controll = []
    error_true_neg_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            total_tracks = len(text_track_array[j]) + len(false_positive_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            error_true_neg = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/total_tracks)*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
                        
            error_true_neg_image.append(error_true_neg)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total = mean_true_neg + mean_negative + mean_positive
        
        controll.append(total)
        error_true_neg_thresh.append(mean_true_neg)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    percent_function = np.zeros((len(thresh_list)))
    for i in range(len(thresh_list)):
        percent_function[i] = (error_true_neg_thresh[i]*(100-error_negative_thresh[i])*(100-error_positive_thresh[i])/1000000)*100
    max_percent = np.max(percent_function)
    index_max = np.where(percent_function == max_percent)[0][0]
    max_thresh = thresh_list[index_max]
    
    true_max = np.max(error_true_neg_thresh)
    index_true_max = np.where(error_true_neg_thresh == true_max)[0][0]
    true_max_thresh = thresh_list[index_true_max]
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    fig.suptitle("True, false negative and false positive tracks percentage", y=1.05)

    ax.plot(thresh_list, error_true_neg_thresh, color='g', label="True tracks")
    ax.plot(thresh_list, error_negative_thresh, color='b', label="False negative")
    ax.plot(thresh_list, error_positive_thresh, color='r', label="False positive")
    ax.plot(thresh_list, controll, color='c', label="Controll percentage")
    ax.plot(thresh_list, percent_function, color='k', label="Product percent")
    ax.set_xlim((np.min(thresh_list), np.max(thresh_list)))
    ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Threshold")
    ax.set_ylabel("Percentage")
    ax.legend() 
    plt.tight_layout()
    #plt.savefig("all_track_{}".format(number), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return max_thresh, true_max_thresh
```
