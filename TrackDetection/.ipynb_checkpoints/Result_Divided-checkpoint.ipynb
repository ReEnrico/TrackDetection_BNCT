```python
def get_all_needed(folder_name: str):
    name_list = ["{:03d}".format(i) for i in range(240)]
    lenght_1 = list(range(25, 100, 25))
    name_divided = []
    image_divided = []
    true_text_divided = []
    algo_text_divided = []
    lenght_divided = []
    
    temp_name = []
    temp_image = []
    temp_text = []
    temp_algo = []
    for name in name_list:
        text = np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name)))
        if len(text) == 0:
            temp_name.append(name)
            temp_image.append(io.imread(os.path.join("..", "img",
                                            "{}.jpg".format(name))))
            temp_text.append(text)
            temp_algo.append(np.loadtxt(os.path.join(folder_name,
                                          "{}.txt".format(name))))
    name_divided.append(temp_name)
    image_divided.append(temp_image)
    true_text_divided.append(temp_text)
    algo_text_divided.append(temp_algo)
    lenght_divided.append([str(0), str(0)])
    
    temp_name = []
    temp_image = []
    temp_text = []
    temp_algo = []
    for name in name_list:
        text = np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name)))
        if len(text) >= 1 and len(text) < 25:
            temp_name.append(name)
            temp_image.append(io.imread(os.path.join("..", "img",
                                            "{}.jpg".format(name))))
            temp_text.append(text)
            temp_algo.append(np.loadtxt(os.path.join(folder_name,
                                          "{}.txt".format(name))))
    name_divided.append(temp_name)
    image_divided.append(temp_image)
    true_text_divided.append(temp_text)
    algo_text_divided.append(temp_algo)
    lenght_divided.append([str(1), str(25)])
    
    for indx in range(1, len(lenght_1)):
        temp_name = []
        temp_image = []
        temp_text = []
        temp_algo = []
        check = False
        for i in range(len(name_list)):
            text = np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name_list[i])))
            if len(text) >= lenght_1[indx-1] and len(text) < lenght_1[indx]:
                temp_name.append(name_list[i])
                temp_image.append(io.imread(os.path.join("..", "img",
                                            "{}.jpg".format(name_list[i]))))
                temp_text.append(text)
                temp_algo.append(np.loadtxt(os.path.join(folder_name,
                                          "{}.txt".format(name_list[i]))))               
                check = True
        if check:
            name_divided.append(temp_name)
            image_divided.append(temp_image)
            true_text_divided.append(temp_text)
            algo_text_divided.append(temp_algo)
            lenght_divided.append([str(lenght_1[indx-1]), str(lenght_1[indx])])
            
    lenght_2 = list(range(550, 700, 25))      
    for indx in range(1, len(lenght_2)):
        temp_name = []
        temp_image = []
        temp_text = []
        temp_algo = []
        check = False
        for i in range(len(name_list)):
            text = np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name_list[i])))
            if len(text) >= lenght_2[indx-1] and len(text) < lenght_2[indx]:
                temp_name.append(name_list[i])
                temp_image.append(io.imread(os.path.join("..", "img",
                                            "{}.jpg".format(name_list[i]))))
                temp_text.append(text)
                temp_algo.append(np.loadtxt(os.path.join(folder_name,
                                          "{}.txt".format(name_list[i]))))               
                check = True
        if check:
            name_divided.append(temp_name)
            image_divided.append(temp_image)
            true_text_divided.append(temp_text)
            algo_text_divided.append(temp_algo)
            lenght_divided.append([str(lenght_2[indx-1]), str(lenght_2[indx])])

    return name_divided, image_divided, true_text_divided, algo_text_divided, lenght_divided, 


def false_negative(track_array: np.array, 
                   true_array: np.array):
    
    array_true = []
    array_false_negative = []
    # loop over the images
    for i in range(len(track_array)):
        true = []
        false_negative = []
        if len(true_array[i]) != 0:
            if not np.isscalar(true_array[i][0]):
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto e con più di un blob
                        algo_index = [k for k in range(len(track_array[i]))]
                        for blob in true_array[i]:
                            y, x, r = blob
                            check = False
                            for a_i in algo_index:
                                y_a, x_a, r_a = track_array[i][a_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(track_array[i][a_i])
                                    algo_index.remove(a_i)
                                    break
                            if not check:
                                false_negative.append(blob)
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        true_index = [k for k in range(len(true_array[i]))]
                        for t_i in true_index:
                            y, x, r = true_array[i][t_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i])
                                true_index.remove(t_i)
                                break
                        for t_i in true_index:
                            false_negative.append(true_array[t_i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                else:
                    # immagine non vuota e con più di un blob
                    # file algoritmo vuoto
                    for blob in true_array[i]:
                        false_negative.append(blob)
                    array_true.append(true)
                    array_false_negative.append(false_negative)
                        
            else:
                y, x, r = true_array[i]
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto con più di un blob
                        check = False
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(a_blob)
                                check = True
                                break
                        if not check:
                            false_negative.append(true_array[i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                            
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto e con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_negative.append(true_array[i])
                        array_true.append(true)
                        array_false_negative.append(false_negative)  
                            
                else:
                    # immagine non vuota e con un solo blob
                    # file algoritmo vuoto
                    false_negative.append(true_array[i])
                    array_false_negative.append(false_negative)
                        
        else:
            # immagine vuota
            # file algoritmo non importante
            array_true.append(true)
            array_false_negative.append(false_negative)
                
    return array_true, array_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    
    
    array_true = []
    array_false_positive = []
    # loop over the images
    for i in range(len(track_array)):
        true = []
        false_positive = []
        if len(true_array[i]) != 0:
            if not np.isscalar(true_array[i][0]):
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto e con più di un blob
                        true_index = [k for k in range(len(true_array[i]))]
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            check = False
                            for t_i in true_index:
                                y, x, r = true_array[i][t_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(a_blob)
                                    true_index.remove(t_i)
                                    break
                            if not check:
                                false_positive.append(a_blob)
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo non vuoto con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        check = False
                        for blob in true_array[i]:
                            y, x, r = blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i])
                                check = True
                                break
                        if not check:
                            false_positive.append(track_array[i])
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                            
                else:
                    # immagine non vuota e con più di un blob
                    # file algoritmo vuoto
                    array_true.append(true)
                    array_false_positive.append(false_positive)
                        
            else:
                y, x, r = true_array[i]
                if len(track_array[i]) != 0:
                    if not np.isscalar(track_array[i][0]):
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuota e con più di un blob
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(a_blob)
                            else:
                                false_positive.append(a_blob)
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                           
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo non vuoto e con un solo blob
                        y_a, x_a, r_a = track_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_positive.append(track_array[i])
                        array_true.append(true)
                        array_false_positive.append(false_positive)  
                            
                else:
                    # immagine non vuota e con un solo blob
                    # file algoritmo vuoto
                    array_true.append(true)
                    array_false_positive.append(false_positive)
                        
        else:
            array_true.append(true)
            if len(track_array[i]) != 0:
                if not np.isscalar(track_array[i][0]):
                    # immagine vuota
                    # file algortimo non vuoto e con più di un blob
                    for a_blob in track_array[i]:
                        false_positive.append(a_blob)
                else:
                    # immagine vuota
                    # file algoritmo non vuoto e con un blob
                    false_positive.append(track_array[i])
            array_false_positive.append(false_positive)
        
    return array_true, array_false_positive

def get_mean(text_track_array: np.array,
             true_track_array: np.array,
             false_negative_array: np.array,
             false_positive_array: np.array):

    true_track = []
    false_negative_track = []
    false_positive_track = []
    for i in range(len(text_track_array)):
        false_negative = len(false_negative_array[i])
        false_positive = len(false_positive_array[i])
        true_found = len(true_track_array[i])
        total_tracks = len(text_track_array[i]) + false_positive
        
        if len(text_track_array[i]) != 0:
            error_true_neg = (true_found/total_tracks)*100
            error_negative = (false_negative/total_tracks)*100
            error_positive = (false_positive/total_tracks)*100
        
        else:
            if false_positive != 0:
                error_true_neg = 100 - (false_positive/total_tracks)*100
                error_negative = false_negative
                error_positive = (false_positive/total_tracks)*100
                
            else:
                error_true_neg = 100
                error_negative = false_negative
                error_positive = false_positive
    
        true_track.append(error_true_neg)
        false_negative_track.append(error_negative)
        false_positive_track.append(error_positive)
    
    median = [np.mean(true_track),
              np.mean(false_negative_track),
              np.mean(false_positive_track)]
    return median

def histogram(group_mean: list, lenght_list: list):
    true_value = [num[0] for num in group_mean]
    negative_value = [num[1] for num in group_mean]
    positive_value = [num[2] for num in group_mean]
    mean = [round(np.mean(true_value), 2),
            round(np.mean(negative_value), 2), 
            round(np.mean(positive_value), 2)]
    #for i in range(len(true_value)):
    #    print(round((true_value[i]), 2), round((negative_value[i]), 2), round((positive_value[i]), 2))
    bottom_negative = true_value
    bottom_positive = [num1+num2 for num1, num2 in zip(true_value, negative_value)]
    x_value = list(range(9))
    x_label = ["0"]
    for string in lenght_list[1:]:
        x_label.append("{}-{}".format(string[0], string[1]))
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 7))
    fig.suptitle("Percentage of true, false negative, true track")
    p1 = ax.bar(x_value, true_value, color="green", label="True")
    p2 = ax.bar(x_value, negative_value, bottom=bottom_negative, 
                     color="blue", label="False Negative")
    p3 = ax.bar(x_value, positive_value, bottom=bottom_positive,
                     color="red", label="False Positive")  
    ax.set_xticks(x_value)
    ax.set_xticklabels(x_label)
    ax.set_xlabel("Number of tracks for images group")
    ax.set_ylabel("Percent value")
    
    plt.legend()
    plt.savefig("lg_tentativo_4", bbox_inches='tight')
    plt.show()
    plt.close()
    return mean

def plot_group(image_array: list,
               text_array: list,
               track_array: list,
               false_negative_array: list,
               false_positive_array: list,
               name_list: list,
               image_index: int,
               group_index: int):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 11))
    fig.suptitle("Image {}".format(name_list[group_index][image_index]), y=0.8)

    ax1.imshow(image_array[group_index][image_index], cmap='gray')
    if len(text_array[group_index][image_index]) != 0:
        if not np.isscalar(text_array[group_index][image_index][0]):
            ax1.scatter(text_array[group_index][image_index][:, 1],
                        text_array[group_index][image_index][:, 0],
                        s=text_array[group_index][image_index][:, 2],
                        color='black',
                        label="True track")
        else:
            ax1.scatter(text_array[group_index][image_index][1],
                        text_array[group_index][image_index][0],
                        s=text_array[group_index][image_index][2],
                        color='k',
                        label="True track")

    if len(track_array[group_index][image_index]) != 0:        
        if not np.isscalar(track_array[group_index][image_index][0]):
            ax1.scatter(track_array[group_index][image_index][:, 1],
                        track_array[group_index][image_index][:, 0],
                        s=track_array[group_index][image_index][:, 2],
                        color='y',
                        label="Algorithm track")
        else:
            ax1.scatter(track_array[group_index][image_index][1],
                        track_array[group_index][image_index][0],
                        s=track_array[group_index][image_index][2],
                        color='y',
                        label="Algorithm track")
    ax1.set_title("True {} \nFounded {}".format(len(text_array[group_index][image_index]),
                                                len(track_array[group_index][image_index])))
    ax1.legend()
    
    ax2.imshow(image_array[group_index][image_index], cmap='gray')
    if len(false_negative_array[group_index][image_index]) != 0:
        if not np.isscalar(false_negative_array[group_index][image_index][0]):
            ax2.scatter(np.asarray(false_negative_array[group_index][image_index])[:, 1],
                        np.asarray(false_negative_array[group_index][image_index])[:, 0],
                        s=np.asarray(false_negative_array[group_index][image_index])[:, 2],
                        color='b',
                        label="False negative track")
        else:
            ax2.scatter(false_negative_array[group_index][image_index][1],
                        false_negative_array[group_index][image_index][0],
                        s=false_negative_array[group_index][image_index][2],
                        color='b',
                        label="False negative track")
    
    if len(false_positive_array[group_index][image_index]) != 0:
        if not np.isscalar(false_positive_total[group_index][image_index][0]):
            ax2.scatter(np.asarray(false_positive_array[group_index][image_index])[:, 1],
                        np.asarray(false_positive_array[group_index][image_index])[:, 0],
                        s=np.asarray(false_positive_array[group_index][image_index])[:, 2],
                        color='r',
                        label="False positive track")
        else:
            ax2.scatter(false_positive_array[group_index][image_index][1],
                        false_positive_array[group_index][image_index][0],
                        s=false_positive_array[group_index][image_index][2],
                        color='r',
                        label="False positive track")
    ax2.set_title("False negative {} \nFalse positive {}".format(len(false_negative_array[group_index][image_index]),
                                                                 len(false_positive_array[group_index][image_index])))
    ax2.legend()
    
    plt.show()
    plt.close()
```

# LOG ALGORITHM
First è semplice analisi con min sigma 3, max sigma 20, num sigma 10, threshold 0.0254. 

Second ho abbassato threshold a 0.0190, sono diminuiti di falsi negativi ma aumentati i falsi positivi. 

Come terzo tentativo provo a fare num sigma 20 il resto uguale; vi sono più kernel applicati alle immagini, più punti che creano il "path" dei blob lungo lo scale-space, ne risulta che le percentuali di falsi negativi sono dimunuite, quindi influiesce sulla sensibilità, ma aumenta anche lo sporco.

Quarto tentativo: provo a tenere num sigma a 10 ma aumento il raggio minimo di 1 e dimezzo il raggio massimo. Sono leggermente diminuiti i falsi positivi e contemporaneamente aumentati i falsi negativi dato che sono state tagliate via le tracce più piccole. 

Quinto tentativo: come sopra ma min sigma 3.5, guadagno in falsi negativi ma aumentano i falsi positivi.

Sesto: dimezzo num sigma rispetto all'iniziale, rimetto min sigma 3 e max sigma a 20, confrontare con i secondi dati. I falsi negativi sono leggermente aumentati.

```python
def plot_image(text: np.array,
               algo_track: np.array,
               algo_false_negative_track: np.array,
               algo_false_positive_track: np.array,               
               image: np.array,
               name: str):
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 15))
    fig.suptitle("Image {}".format(name))
    
    ax1.imshow(image, cmap="gray")
    if len(text) != 0:
        ax1.scatter(text[:, 1],
                    text[:, 0],
                    s=text[:, 2],
                    color='y',
                    alpha=0.30,
                    label="True")
    if len(algo_track) != 0:
        if not np.isscalar(algo_track[0]):
            ax1.scatter(algo_track[:, 1],
                        algo_track[:, 0],
                        s=algo_track[:, 2],
                        color='c',
                        alpha=0.30,
                        label="Algorithm")
        else:
            ax1.scatter(algo_track[1],
                        algo_track[0],
                        s=algo_track[2],
                        color='c',
                        alpha=0.30,
                        label="Algorithm")
        ax1.legend()
    ax1.set_title("True tracks {} and\n algorith tracks {}".format(len(text),
                                                               len(algo_track)))


    ax2.imshow(image, cmap="gray")
    if len(algo_false_negative_track) != 0:
        if np.isscalar(algo_false_negative_track[0]):
             ax2.scatter(algo_false_negative_track[1],
                         algo_false_negative_track[0],
                         s=algo_false_negative_track[2],
                         color='b',
                         label="False Negative")
        else:
            ax2.scatter(np.array(algo_false_negative_track)[:, 1],
                        np.array(algo_false_negative_track)[:, 0],
                        s=np.array(algo_false_negative_track)[:, 2],
                        color='b',
                        label="False Negative")
           
    if len(algo_false_positive_track) != 0:
        if not np.isscalar(algo_false_positive_track[0]):
            for blob in np.array(algo_false_positive_track):
                y, x, r = blob
                c = plt.Circle((x, y), r, color='r', fill=False)
                ax2.add_patch(c)
        else:
            y, x, r = np.array(algo_false_positive_track)
            c = plt.Circle((x, y), r, color='r', fill=False)
            ax2.add_patch(c)
    ax2.set_title("False positive {p_num} and\n false negative {n_num}".format(p_num=len(algo_false_positive_track),
                                                                               n_num=len(algo_false_negative_track)))

    plt.show()
    plt.close()

for j in range(len(names[i])):
        plot_image(texts[i][j],
                   log_track[i][j],
                   log_false_negative_track[j],
                   log_false_positive_track[j],               
                   images[i][j],
                   names[i][j])
```
