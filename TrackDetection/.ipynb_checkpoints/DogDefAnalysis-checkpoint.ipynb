```python
import os
import glob

import matplotlib.pyplot as plt
import numpy as np
from skimage import io
```

```python
def upload_main_file(name_array: list):
    image_array = []
    text_array = []
    lenght_array = []
    for name in name_array:
        img = io.imread(os.path.join("..", "img", "{}.jpg".format(name)))
        image_array.append(img)
        txt = np.loadtxt(os.path.join("..", "img", "{}.txt".format(name)))
        text_array.append(txt)
        lnght = len(txt)
        lenght_array.append(lnght)
        
    return image_array, text_array, lenght_array


def get_file_name(name_array: list, algorithm_name: str, directory_name: str):
    alg_file = []
    for name in name_array:
        alg = glob.glob(os.path.join("{algo}_{dir}".format(algo=algorithm_name, dir=directory_name), 
                                     "{name}_*.txt".format(name=name)))
        alg.sort()
        alg_file.append(alg)

    return alg_file


def get_tracks(name_array):
    track_array = []
    for i in range(len(name_array)):
        tmp = []
        for j in range(len(name_array[i])):
            dmp = np.loadtxt(name_array[i][j])
            tmp.append(dmp)
        track_array.append(tmp)
        
    return np.asarray(track_array, dtype=object)
```

```python
names = ["{:03d}".format(i) for i in range(240)]
images, texts, lenghts = upload_main_file(names)
dog25_txt = get_file_name(names, "Raggio", "25")
dog25_track = get_tracks(dog25_txt)
```

```python
def false_negative_max(track_array: np.array, 
                   true_array: np.array,  
                   name_list: list):
    global_information = []
    total_true = []
    total_false_negative = []
    # loop over images
    for i in range(len(true_array)):
        if len(true_array[i]) != 0:
            array_true = []
            array_false_negative = []
            a_blob_j = [j for j in range(len(track_array[i][0]))]
            for blob in true_array[i]:
                y, x, r, = blob
                check = False                
                for a_j in a_blob_j:
                    y_a, x_a, r_a = track_array[i][0][a_j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    #r_max = np.array([r, r_a]).max()
                    if d2 <= (r)**2:
                        check = True
                        break
                if check:
                    array_true.append(track_array[i][0][a_j])
                    a_blob_j.remove(a_j)
                else:
                    array_false_negative.append(blob)
            global_information.append([name_list[i], len(array_true), len(array_false_negative)])
            total_true.append(array_true)
            total_false_negative.append(np.array(array_false_negative, dtype=object))
        else:
            global_information.append([name_list[i], 0, 0])
            total_true.append([])
            total_false_negative.append([])
            
    return global_information, np.array(total_true, dtype=object), np.array(total_false_negative, dtype=object)

def false_positive_max(track_array: np.array, 
                   true_array: np.array, 
                   name_list: list):
    
    global_information = []
    total_true = []
    total_false_positive = []
    # loop over images
    for i in range(len(true_array)):
        if len(true_array[i]) != 0:
            array_true = []
            array_false_positive = []
            blob_j = [j for j in range(len(true_array[i]))]
            for a_blob in track_array[i][0]:
                y_a, x_a, r_a = a_blob
                check = False
                for b_j in blob_j:
                    y, x, r, = true_array[i][b_j]
                    d2 = (x - x_a)**2 + (y - y_a)**2
                    #r_max = np.array([r, r_a]).max()
                    if d2 <= (r)**2:
                        check = True
                        array_true.append(a_blob)
                        blob_j.remove(b_j)
                        break
                if not check:
                    array_false_positive.append(a_blob)
            
            global_information.append([name_list[i], len(array_true), len(array_false_positive)])
            total_true.append(array_true)
            total_false_positive.append(array_false_positive)
        else:
            if len(track_array[i]) != 0:
                array_false_positive = [blob for blob in track_array[i]]
                total_false_positive.append(array_false_positive)
                global_information.append([name_list[i], 0, len(array_false_positive)])
            else:
                global_information.append([name_list[i], 0, 0])
                total_true.append([])
                total_false_positive.append([])
                
    return global_information, total_true, total_false_positive
```

```python
thresh_false_neg, n_true_track, false_negative_track = false_negative_max(dog_track, texts, names)
thresh_false_pos, p_true_track, false_positive_track = false_positive_max(dog_track, texts, names)
```

```python
def show_an_image(true_array: np.array, 
                  algorithm_array: np.array,
                  true_track_array: np.array,
                  false_negative_array: np.array,
                  false_positive_array: np.array,
                  image: np.array,
                  image_name: str):
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, constrained_layout=True, figsize=(12, 9))
    fig.suptitle("Image {}, threshold {}".format(image_name, "0.01"))
    
    ax1.imshow(image, cmap="gray")
    for blob in true_array:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='green', linewidth=0.5, fill=False)
        ax1.add_patch(c)
    ax1.set_title("True tracks")
    
    ax2.imshow(image, cmap="gray")
    for blob in algorithm_array:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='blue', linewidth=0.5, fill=False)
        ax2.add_patch(c)
    ax2.set_title("Algorithm's tracks")
    
    ax3.imshow(image, cmap="gray")
    for blob in true_track_array:
        y, x, r = blob
        c = plt.Circle((x, y), r, color='yellow', linewidth=0.5, fill=False)
        ax3.add_patch(c)
    ax3.set_title("True track found")
    
    ax4.imshow(image, cmap="gray")
    if not len(false_negative_array) == 0:
        if not np.isscalar(false_negative_array[0]):
            for blob in false_negative_array:
                y, x, r = blob
                c = plt.Circle((x, y), r, color='orange', linewidth=0.5, fill=False)
                ax4.add_patch(c)
            for blob in false_positive_array:
                y, x, r = blob
                c = plt.Circle((x, y), r, color='red', linewidth=0.5, fill=False)
                ax4.add_patch(c)
    
    plt.show()
    plt.close()
```
