```python
def get_all_needed(folder_name: str):
    name_list = ["{:03d}".format(i) for i in range(240)]
    image_array = [io.imread(os.path.join("..", "img",
                    "{}.jpg".format(name))) for name in name_list]
    track_array = [np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name))) for name in name_list]
    
    track_file = [np.loadtxt(os.path.join(folder_name,
                                          "{name}_0.0272.txt".format(name=name))) for name in name_list]
    
    return name_list, image_array, track_array, track_file


def false_negative(track_array: np.array, 
                   true_array: np.array):
    array_true = []
    array_false_negative = []
    for i in range(len(track_array)):
        true = []
        false_negative = []
        #image non empty
        if len(true_array[i]) != 0:
            #something found
            if len(track_array[i]) != 0:
                #more than 1 blob found
                if not np.isscalar(track_array[i][0]):
                    a_blob_i = [i for i in range(len(track_array[i]))]
                    if not np.isscalar(true_array[i][0]):
                            for blob in true_array[i]:
                                y, x, r = blob
                                check = False                
                                for a_i in a_blob_i:
                                    y_a, x_a, r_a = track_array[i][a_i]
                                    d2 = (x - x_a)**2 + (y - y_a)**2
                                    if d2 <= (r)**2:
                                        check = True
                                        break
                                if check:
                                    true.append(track_array[i][a_i])
                                    a_blob_i.remove(a_i)
                                else:
                                    false_negative.append(blob)
                    else:
                        y, x, r = true_array[i]
                        check = False                
                        for a_i in a_blob_i:
                            y_a, x_a, r_a = track_array[i][a_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                break
                        if check:
                            true.append(track_array[i][a_i])
                            a_blob_i.remove(a_i)
                        else:
                            false_negative.append(blob)

                else:
                    if not np.isscalar(true_array[i][0]):
                        true_i = [i for i in range(len(true_array[i]))]
                        for t_i in true_i:
                            y, x, r = true_array[i][t_i]
                            y_a, x_a, r_a = track_array[i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i])
                                true_i.remove(t_i)
                                break
                        for t_i in true_i:
                            false_negative.append(true_array[i][t_i])
                    else:
                        y, x, r = true_array[i]
                        y_a, x_a, r_a = track_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_negative.append(true_array[i])

                array_true.append(true)
                array_false_negative.append(false_negative)

        else:
            array_true.append(true_array[i])
            array_false_negative.append(false_negative)
            
    return array_true, array_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    array_true = []
    array_false_positive = []
    for i in range(len(track_array)):
        true = []
        false_positive = []
        if len(true_array[i]) != 0:
            if len(track_array[i]) != 0:
                if not np.isscalar(track_array[i][0]):
                    if not np.isscalar(true_array[i][0]):
                        blob_j = [j for j in range(len(true_array[i]))]
                        for a_blob in track_array[i]:
                            y_a, x_a, r_a = a_blob
                            check = False
                            for b_i in blob_j:
                                y, x, r = true_array[i][b_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    check = True
                                    true.append(a_blob)
                                    blob_j.remove(b_i)
                                    break
                            if not check:
                                false_positive.append(a_blob)
                    else:
                        a_blob_i = [i for i in range(len(track_array[i]))]
                        for a_i in a_blob_i:
                            y_a, x_a, r_a = track_array[i][a_i]
                            y, x, r = true_array[i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i][a_i])
                                a_blob_i.remove(a_i)
                                break
                        for a_i in a_blob_i:
                            false_positive.append(track_array[i][a_i])                       

                else:
                    if not np.isscalar(true_array[i][0]):
                        blob_j = [j for j in range(len(true_array[i]))]
                        y_a, x_a, r_a = track_array[i]
                        check = False
                        for b_i in blob_j:
                            y, x, r = true_array[i][b_i]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                check = True
                                true.append(track_array[i])
                                blob_j.remove(b_i)
                                break
                        if not check:
                            false_positive.append(track_array[i])
                    else:
                        y_a, x_a, r_a = track_array[i]
                        y, x, r = true_array[i]
                        d2 = (x - x_a)**2 + (y - y_a)**2
                        if d2 <= (r)**2:
                            true.append(track_array[i])
                        else:
                            false_positive.append(track_array[i])

                array_true.append(true)
                array_false_positive.append(false_positive)
        else:
            array_true.append(true_array[i])
            array_false_positive.append(track_array[i])
     
    return array_true, array_false_positive

def number_track_plot(text_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      name_list: list,
                      algo_name: str):

    controll = []
    true_track = []
    false_negative_track = []
    false_positive_track = []
    for i in range(len(name_list)):
        false_negative = len(false_negative_array[i])
        false_positive = len(false_positive_array[i])
        true_found = len(true_track_array[i])
        total_tracks = len(text_track_array[i]) + false_positive
        
        if total_tracks != 0:
            error_true_neg = np.round((true_found/total_tracks)*100, 2)
            error_negative = np.round((false_negative/total_tracks)*100, 2)
            error_positive = np.round((false_positive/total_tracks)*100, 2)
            total = error_true_neg + error_negative + error_positive
        
        else:
            error_true_neg = 0
            error_negative = 0
            error_positive = 0
            total = 100          
            
        controll.append(total)
        true_track.append(error_true_neg)
        false_negative_track.append(error_negative)
        false_positive_track.append(error_positive)
    
    mean_true = np.mean(np.asarray(true_track))
    std_true = np.std(np.asarray(true_track))
    mean_positive = np.mean(np.asarray(false_positive_track))
    std_positive = np.std(np.asarray(false_positive_track))
    mean_negative = np.mean(np.asarray(false_negative_track))
    std_negative = np.std(np.asarray(false_negative_track))
    name = [int(name) for name in name_list]

    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    fig.suptitle("{} true, false negative and false positive tracks percentage per image".format(algo_name), y=1.05)

    ax.plot(name, true_track, color='b', label="True tracks")
    ax.plot(name, false_negative_track, color='g', label="False negative")
    ax.plot(name, false_positive_track, color='r', label="False positive")
    ax.plot(name, controll, color='k', label="Controll percentage")
    ax.set_xlim((np.min(name), np.max(name)))
    #ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("Images")
    ax.set_ylabel("Percentage")
    ax.set_title("""True tracks mean percentage {mt:.2f} +- {st:.2f} 
    False negative mean percentage {mf:.2f} +- {sf:.2f}
    False positive mean percentage {mp:.2f} +- {sp:.2f}""".format(mt=mean_true,
                                                                  st=std_true,
                                                                  mf=mean_negative,
                                                                  sf=std_negative,
                                                                  mp=mean_positive,
                                                                  sp=std_positive))
    #ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.legend() 
    plt.tight_layout()
    #plt.savefig("All_images_{}".format(algo_name), bbox_inches='tight')
    #plt.show()
    plt.close()

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))
    fig.suptitle("{} histogram".format(algo_name), y=1.05)

    ax1.hist(true_track, bins=15, color='b', label="True tracks")
    ax1.axvline(np.median(true_track), color="black", linestyle="--", alpha=0.25)
    ax1.set_xlim((0, 100))
    ax1.set_xlabel("Percentage")
    ax1.set_ylabel("Number of images")
    ax1.set_title("True tracks mean {:.2f} +- {:.2f}".format(np.median(true_track),
                                                             np.std(true_track)), y=1.05)
    ax1.legend()
    
    ax2.hist(false_negative_track, bins=10, color='g', label="False negative")
    ax2.axvline(np.median(false_negative_track), color="black", linestyle="--", alpha=0.25)
    ax2.set_xlim((0, 100))
    ax2.set_xlabel("Percentage")
    ax2.set_ylabel("Number of images")
    ax2.set_title("False negative tracks mean {:.2f} +- {:.2f}".format(np.median(false_negative_track),
                                                                       np.std(false_negative_track)), y=1.05)
    ax2.legend()
    
    ax3.hist(false_positive_track, bins=15, color='r', label="False positive")
    ax3.axvline(np.median(false_positive_track), color="black", linestyle="--", alpha=0.25)
    ax3.set_xlim((0, 100))
    ax3.set_xlabel("Percentage")
    ax3.set_ylabel("Number of images")
    ax3.set_title("False positive {:.2f} +- {:.2f}".format(np.median(false_positive_track),
                                                           np.std(false_positive_track)), y=1.05)
    ax3.legend()
    
    plt.tight_layout()
    #plt.savefig("First_{}".format(algo_name), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return true_track, false_negative_track, false_positive_track
```

```python
def get_all_file(folder_name: str):
    name_list = ["{:03d}".format(i) for i in range(240)]
    image_array = [io.imread(os.path.join("..", "img",
                    "{}.jpg".format(name))) for name in name_list]
    true_array = [np.loadtxt(os.path.join("..", "img",
                            "{}_half.txt".format(name))) for name in name_list]
    
    if "Max_Sigma" in folder_name:
        temp_name = glob.glob(os.path.join(folder_name, "000_*.txt"))
        temp_name.sort()
        variable = [file[-6:-4] for file in temp_name]
        
    if "Max_Sigma2" in folder_name:
        temp_name = glob.glob(os.path.join(folder_name, "000_*.txt"))
        temp_name.sort()
        variable = [file[-6:-4] for file in temp_name]
        
    if "Min_Sigma" in folder_name:
        temp_name = glob.glob(os.path.join(folder_name, "000_*.txt"))
        temp_name.sort()
        variable = [file[-7:-4] for file in temp_name]
    
    if "Sigma_Ratio" in folder_name:
        temp_name = glob.glob(os.path.join(folder_name, "000_*.txt"))
        temp_name.sort()
        variable = [file[-7:-4] for file in temp_name]
        
    track_array = []
    for val in variable:
        image_list = []
        for name in name_list:
            text = np.loadtxt(os.path.join(folder_name, "{}_{}.txt".format(name, val)))
            image_list.append(text)
        track_array.append(image_list)
    
    return name_list, image_array, true_array, variable, track_array

def false_negative(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_negative = []
    # loop over radius value
    for i in range(len(track_array)):
        array_true = []
        array_false_negative = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_negative = []
            if len(true_array[j]) != 0:
                if not np.isscalar(true_array[j][0]):
                    if len(track_array[i][j]) != 0:
                        if not np.isscalar(track_array[i][j][0]):
                            # immagine non vuota e con più di un blob
                            # file algoritmo non vuoto e con più di un blob
                            algo_index = [k for k in range(len(track_array[i][j]))]
                            for blob in true_array[j]:
                                y, x, r = blob
                                check = False
                                for a_i in algo_index:
                                    y_a, x_a, r_a = track_array[i][j][a_i]
                                    d2 = (x - x_a)**2 + (y - y_a)**2
                                    if d2 <= (r)**2:
                                        check = True
                                        true.append(track_array[i][j][a_i])
                                        algo_index.remove(a_i)
                                        break
                                if not check:
                                    false_negative.append(blob)
                            array_true.append(true)
                            array_false_negative.append(false_negative)
                            
                        else:
                            # immagine non vuota e con più di un blob
                            # file algoritmo non vuoto con un solo blob
                            y_a, x_a, r_a = track_array[i][j]
                            true_index = [k for k in range(len(true_array[j]))]
                            for t_i in true_index:
                                y, x, r = true_array[j][t_i]
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    true.append(track_array[i][j])
                                    true_index.remove(t_i)
                                    break
                            for t_i in true_index:
                                false_negative.append(true_array[j][t_i])
                            array_true.append(true)
                            array_false_negative.append(false_negative)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo vuoto
                        for blob in true_array[j]:
                            false_negative.append(blob)
                        array_true.append(true)
                        array_false_negative.append(false_negative)
                        
                else:
                    y, x, r = true_array[j]
                    if len(track_array[i][j]) != 0:
                        if not np.isscalar(track_array[i][j][0]):
                            # immagine non vuota e con un solo blob
                            # file algoritmo non vuoto con più di un blob
                            check = False
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    true.append(a_blob)
                                    check = True
                                    break
                            if not check:
                                false_negative.append(true_array[j])
                            array_true.append(true)
                            array_false_negative.append(false_negative)
                            
                        else:
                            # immagine non vuota e con un solo blob
                            # file algoritmo non vuoto e con un solo blob
                            y_a, x_a, r_a = track_array[i][j]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i][j])
                            else:
                                false_negative.append(true_array[j])
                            array_true.append(true)
                            array_false_negative.append(false_negative)  
                            
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo vuoto
                        false_negative.append(true_array[j])
                        array_false_negative.append(false_negative)
                        
            else:
                # immagine vuota
                # file algoritmo non importante
                array_true.append(true)
                array_false_negative.append(false_negative)
                
        total_true.append(array_true)
        total_false_negative.append(array_false_negative)
    return total_true, total_false_negative


def false_positive(track_array: np.array, 
                   true_array: np.array):
    total_true = []
    total_false_positive = []
    # loop over radius value
    for i in range(len(track_array)):
        array_true = []
        array_false_positive = []
        # loop over the images
        for j in range(len(track_array[i])):
            true = []
            false_positive = []
            if len(true_array[j]) != 0:
                if not np.isscalar(true_array[j][0]):
                    if len(track_array[i][j]) != 0:
                        if not np.isscalar(track_array[i][j][0]):
                            # immagine non vuota e con più di un blob
                            # file algoritmo non vuoto e con più di un blob
                            true_index = [k for k in range(len(true_array[j]))]
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                check = False
                                for t_i in true_index:
                                    y, x, r = true_array[j][t_i]
                                    d2 = (x - x_a)**2 + (y - y_a)**2
                                    if d2 <= (r)**2:
                                        check = True
                                        true.append(a_blob)
                                        true_index.remove(t_i)
                                        break
                                if not check:
                                    false_positive.append(a_blob)
                            array_true.append(true)
                            array_false_positive.append(false_positive)
                            
                        else:
                            # immagine non vuota e con più di un blob
                            # file algoritmo non vuoto con un solo blob
                            y_a, x_a, r_a = track_array[i][j]
                            check = False
                            for blob in true_array[j]:
                                y, x, r = blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    true.append(track_array[i][j])
                                    check = True
                                    break
                            if not check:
                                false_positive.append(track_array[i][j])
                            array_true.append(true)
                            array_false_positive.append(false_positive)
                            
                    else:
                        # immagine non vuota e con più di un blob
                        # file algoritmo vuoto
                        array_true.append(true)
                        array_false_negative.append(false_positive)
                        
                else:
                    y, x, r = true_array[j]
                    if len(track_array[i][j]) != 0:
                        if not np.isscalar(track_array[i][j][0]):
                            # immagine non vuota e con un solo blob
                            # file algoritmo non vuota e con più di un blob
                            for a_blob in track_array[i][j]:
                                y_a, x_a, r_a = a_blob
                                d2 = (x - x_a)**2 + (y - y_a)**2
                                if d2 <= (r)**2:
                                    true.append(a_blob)
                                else:
                                    false_positive.append(a_blob)
                            array_true.append(true)
                            array_false_positive.append(false_positive)
                            
                        else:
                            # immagine non vuota e con un solo blob
                            # file algoritmo non vuoto e con un solo blob
                            y_a, x_a, r_a = track_array[i][j]
                            d2 = (x - x_a)**2 + (y - y_a)**2
                            if d2 <= (r)**2:
                                true.append(track_array[i][j])
                            else:
                                false_positive.append(track_array[i][j])
                            array_true.append(true)
                            array_false_positive.append(false_positive)  
                            
                    else:
                        # immagine non vuota e con un solo blob
                        # file algoritmo vuoto
                        array_true.append(true)
                        array_false_positive.append(false_positive)
                        
            else:
                array_true.append(true)
                if len(track_array[i][j]) != 0:
                    if not np.isscalar(track_array[i][j][0]):
                        # immagine vuota
                        # file algortimo non vuoto e con più di un blob
                        for a_blob in track_array[i][j]:
                            false_positive.append(a_blob)
                    else:
                        false_positive.append(track_array[i][j])
                array_false_positive.append(false_positive)
        total_true.append(array_true)
        total_false_positive.append(array_false_positive)
    return total_true, total_false_positive

def global_plot(true_track_array: np.array,
                false_negative_array: np.array, 
                false_positive_array: np.array,
                variable_list: list,
                algo_name: str,
                variable_name: str):
    # analisi globale usando la media dei risultati    
    true_mean_radius = []
    for i in range(len(true_track_array)):
        mean_image = []
        for j in range(len(true_track_array[i])):
            if len(true_track_array[i][j]):
                mean_image.append(len(true_track_array[i][j]))
        mean = int(np.mean(mean_image))
        true_mean_radius.append(mean)
    
    negative_mean_radius = []
    for i in range(len(false_negative_array)):
        mean_image = []
        for j in range(len(false_negative_array[i])):
            mean_image.append(len(false_negative_array[i][j]))
        mean = int(np.mean(mean_image))
        negative_mean_radius.append(mean)

    positive_mean_radius = []
    for i in range(len(false_positive_array)):
        mean_image = []
        for j in range(len(false_positive_array[i])):
            mean_image.append(len(false_positive_array[i][j]))
        mean = int(np.mean(mean_image))
        positive_mean_radius.append(mean)
    
    #variable_int = [int(val) for val in variable_list]
    bottom_negative = true_mean_radius
    bottom_positive = [num1+num2 for num1, num2 in zip(true_mean_radius, negative_mean_radius)]
    # all data needed     
    
    fig, ax = plt.subplots(1, 1, constrained_layout=True, figsize=(10, 7))
    fig.suptitle("{} algorithm's beheviour".format(algo_name), y=1.05)
    
    p1 = ax.bar(variable_list, true_mean_radius, color="green", label="True")
    p2 = ax.bar(variable_list, negative_mean_radius, bottom=bottom_negative, 
                 color="blue", label="False Negative")
    p3 = ax.bar(variable_list, positive_mean_radius, bottom=bottom_positive,
                 color="red", label="False Positive")  
    #ax.set_xlim((np.min(variable_list), np.max(variable_list)))
    ax.set_xticks(variable_list)
    #ax.set_xticklabels(radius, rotation=65)
    ax.set_xlabel("{} Value".format(variable_name))
    ax.set_ylabel("Number of Tracks")
    ax.set_title("True, false segative, false positive tracks mean in function of {}".format(variable_name))
    #ax.set_xscale("log")
    #ax.set_yscale("log")
    ax.legend()
    #plt.autoscale(enable=True, axis='x', tight=None)
    plt.savefig("istogramma_{}_{}".format(variable_name, algo_name), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return None

def number_track_plot(text_track_array: np.array,
                      true_track_array: np.array,
                      false_negative_array: np.array,
                      false_positive_array: np.array,
                      variable_list: list,
                      algo_name: str,
                      variable_name: str):
    
    
    controll = []
    error_true_neg_thresh = []
    error_negative_thresh = []
    error_positive_thresh = []
    for i in range(len(false_negative_array)):
        error_true_neg_image = []
        error_negative_image = []
        error_positive_image = []
        for j in range(len(false_negative_array[i])):
            total_tracks = len(text_track_array[j]) + len(false_positive_array[i][j])
            true_found   = len(true_track_array[i][j])
            false_negative = len(false_negative_array[i][j])
            false_positive = len(false_positive_array[i][j])
            
            if total_tracks != 0:
                error_true_neg = np.round((true_found/total_tracks)*100, 2)
                error_negative = np.round((false_negative/total_tracks)*100, 2)
                error_positive = np.round((false_positive/total_tracks)*100, 2)
            #else:
            #    error_true_neg = true_found
            #    error_negative = false_negative
            #    error_positive = false_positive
                   
            error_true_neg_image.append(error_true_neg)
            error_negative_image.append(error_negative)
            error_positive_image.append(error_positive)
        
        #error_true_neg_image = [num for num in error_true_neg_image if num != 0]
        #error_negative_image = [num for num in error_negative_image if num != 0]
        #error_positive_image = [num for num in error_positive_image if num != 0]
        mean_true_neg = np.round(np.mean(error_true_neg_image), 2)
        mean_negative = np.round(np.mean(error_negative_image), 2)
        mean_positive = np.round(np.mean(error_positive_image), 2)
        total = mean_true_neg + mean_negative + mean_positive
        
        
        controll.append(total)
        error_true_neg_thresh.append(mean_true_neg)
        error_negative_thresh.append(mean_negative)
        error_positive_thresh.append(mean_positive)
    
    percent_function = np.zeros((len(variable_list)))
    for i in range(len(variable_list)):
        percent_function[i] = (error_true_neg_thresh[i]*(100-error_negative_thresh[i])*(100-error_positive_thresh[i])/1000000)*100
    max_percent = np.max(percent_function)
    index_max = np.where(percent_function == max_percent)[0][0]
    best_value = variable_list[index_max]
    
    true_max = np.max(error_true_neg_thresh)
    index_true_max = np.where(error_true_neg_thresh == true_max)[0][0]
    true_best_value = variable_list[index_true_max]
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    fig.suptitle("""{} true, false negative and false positive tracks percentage
        in function of {}""".format(algo_name, variable_name), y=1.05)

    ax.plot(variable_list, error_true_neg_thresh, color='b', label="True tracks")
    ax.plot(variable_list, error_negative_thresh, color='g', label="False negative")
    ax.plot(variable_list, error_positive_thresh, color='r', label="False positive")
    ax.plot(variable_list, controll, color='c', label="Controll percentage")
    ax.plot(variable_list, percent_function, color='k', label="Product percent")
    #ax.set_xlim((np.min(variable_list), np.max(variable_list)))
    #ax.tick_params(axis='x', labelrotation=65)
    ax.set_xlabel("{}".format(variable_name))
    ax.set_ylabel("Percentage")
    ax.legend() 
    plt.tight_layout()
    plt.savefig("Percentage_{}_{}".format(variable_name, algo_name), bbox_inches='tight')
    plt.show()
    plt.close()
    
    return best_value, true_best_value
```
